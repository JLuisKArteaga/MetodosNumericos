## Valores y Vectores Característicos (Eigenvalores y Eigenvectores)

### Definición

Dada una matriz cuadrada $A$ de tamaño $n \times n$, un **vector característico** (o eigenvector) es un vector no nulo $\mathbf{v} \in \mathbb{R}^n$ que satisface:

$$A\mathbf{v} = \lambda \mathbf{v}$$

donde $\lambda$ es un escalar llamado **valor característico** (o eigenvalor).

### Interpretación Geométrica

Un eigenvector es un vector que, al ser transformado por la matriz $A$, solo cambia de escala (se estira o se comprime) pero **no cambia de dirección** (o apunta en la dirección opuesta si $\lambda < 0$).

El eigenvalor $\lambda$ indica el factor de escala:
- Si $\lambda > 1$: el vector se alarga
- Si $0 < \lambda < 1$: el vector se acorta
- Si $\lambda = 1$: el vector no cambia
- Si $\lambda < 0$: el vector invierte su dirección

### Cálculo de Eigenvalores

Para encontrar los eigenvalores, resolvemos la **ecuación característica**:

$$\det(A - \lambda I) = 0$$

donde $I$ es la matriz identidad de tamaño $n \times n$.

El polinomio resultante $p(\lambda) = \det(A - \lambda I)$ se llama **polinomio característico**, y sus raíces son los eigenvalores de $A$.

### Cálculo de Eigenvectores

Una vez encontrado un eigenvalor $\lambda_i$, el eigenvector asociado $\mathbf{v}_i$ se obtiene resolviendo el sistema:

$$(A - \lambda_i I)\mathbf{v}_i = \mathbf{0}$$

Es decir, el eigenvector pertenece al **espacio nulo** de la matriz $(A - \lambda_i I)$.

### Ejemplo

Para la matriz:

$$A = \begin{pmatrix} 4 & 2 \\ 1 & 3 \end{pmatrix}$$

La ecuación característica es:

$$\det\begin{pmatrix} 4-\lambda & 2 \\ 1 & 3-\lambda \end{pmatrix} = (4-\lambda)(3-\lambda) - 2 = \lambda^2 - 7\lambda + 10 = 0$$

Las raíces son $\lambda_1 = 5$ y $\lambda_2 = 2$.

Para $\lambda_1 = 5$:

$$(A - 5I)\mathbf{v} = \begin{pmatrix} -1 & 2 \\ 1 & -2 \end{pmatrix}\begin{pmatrix} v_1 \\ v_2 \end{pmatrix} = \mathbf{0}$$

Obtenemos $\mathbf{v}_1 = \begin{pmatrix} 2 \\ 1 \end{pmatrix}$ (o cualquier múltiplo escalar).

### Propiedades Importantes

- Una matriz $n \times n$ tiene exactamente $n$ eigenvalores (contando multiplicidades)
- Los eigenvalores de una matriz simétrica son siempre reales
- Los eigenvectores correspondientes a eigenvalores distintos son linealmente independientes
- La **traza** de $A$ (suma de elementos diagonales) es igual a la suma de eigenvalores: $\operatorname{tr}(A) = \sum_{i=1}^n \lambda_i$
- El **determinante** de $A$ es igual al producto de eigenvalores: $\det(A) = \prod_{i=1}^n \lambda_i$

### Aplicaciones

- **Análisis de estabilidad** en sistemas dinámicos
- **Diagonalización** de matrices
- **Análisis de componentes principales (PCA)** en estadística
- **Mecánica cuántica** (valores propios de observables)
- **PageRank** de Google (eigenvector de una matriz de transición)
- **Vibraciones** y modos normales en ingeniería

## Valores y Vectores Característicos (Eigenvalores y Eigenvectores)

### Definición

Dada una matriz cuadrada $A$ de tamaño $n \times n$, un **vector característico** (o eigenvector) es un vector no nulo $\mathbf{v} \in \mathbb{R}^n$ que satisface:

$$A\mathbf{v} = \lambda \mathbf{v}$$

donde $\lambda$ es un escalar llamado **valor característico** (o eigenvalor).

Un eigenvector es un vector que, al ser transformado por la matriz $A$, solo cambia de escala (se estira o se comprime) pero **no cambia de dirección** (o apunta en la dirección opuesta si $\lambda < 0$).

El eigenvalor $\lambda$ indica el factor de escala:
- Si $\lambda > 1$: el vector se alarga (estiramiento)
- Si $0 < \lambda < 1$: el vector se acorta (compresión)
- Si $\lambda = 1$: el vector no cambia
- Si $\lambda < 0$: el vector invierte su dirección

### Relación con la Ley de Hooke

La analogía con **estiramiento y compresión** no es casual. En la **Ley de Hooke** para resortes, la fuerza de restauración es proporcional al desplazamiento:

$$\mathbf{F} = -k\mathbf{x}$$

donde $k$ es la constante del resorte (rigidez) y $\mathbf{x}$ es el desplazamiento desde la posición de equilibrio.

En el contexto de eigenvalores, podemos pensar en la matriz $A$ como un operador que describe un sistema elástico lineal:

| Concepto | Ley de Hooke | Eigenvalores |
|----------|-------------|--------------|
| Constante de rigidez | $k$ | $\lambda$ |
| Desplazamiento | $\mathbf{x}$ | $\mathbf{v}$ |
| Respuesta del sistema | $\mathbf{F} = -k\mathbf{x}$ | $A\mathbf{v} = \lambda\mathbf{v}$ |

En **modos normales de vibración**, los eigenvectores representan las **direcciones principales** de oscilación del sistema, y los eigenvalores están relacionados con las **frecuencias naturales** al cuadrado ($\omega^2$):

$$K\mathbf{v} = \omega^2 M\mathbf{v}$$

donde $K$ es la matriz de rigidez y $M$ es la matriz de masa. Aquí $\lambda = \omega^2$ representa el cuadrado de la frecuencia angular.

### Cálculo de Eigenvalores

Para encontrar los eigenvalores, resolvemos la **ecuación característica**:

$$\det(A - \lambda I) = 0$$

donde $I$ es la matriz identidad de tamaño $n \times n$.

El polinomio resultante $p(\lambda) = \det(A - \lambda I)$ se llama **polinomio característico**, y sus raíces son los eigenvalores de $A$.

### Cálculo de Eigenvectores

Una vez encontrado un eigenvalor $\lambda_i$, el eigenvector asociado $\mathbf{v}_i$ se obtiene resolviendo el sistema:

$$(A - \lambda_i I)\mathbf{v}_i = \mathbf{0}$$

Es decir, el eigenvector pertenece al **espacio nulo** de la matriz $(A - \lambda_i I)$.

### Ejemplo

Para la matriz:

$$A = \begin{pmatrix} 4 & 2 \\ 1 & 3 \end{pmatrix}$$

La ecuación característica es:

$$\det\begin{pmatrix} 4-\lambda & 2 \\ 1 & 3-\lambda \end{pmatrix} = (4-\lambda)(3-\lambda) - 2 = \lambda^2 - 7\lambda + 10 = 0$$

Las raíces son $\lambda_1 = 5$ y $\lambda_2 = 2$.

Para $\lambda_1 = 5$:

$$(A - 5I)\mathbf{v} = \begin{pmatrix} -1 & 2 \\ 1 & -2 \end{pmatrix}\begin{pmatrix} v_1 \\ v_2 \end{pmatrix} = \mathbf{0}$$

Obtenemos $\mathbf{v}_1 = \begin{pmatrix} 2 \\ 1 \end{pmatrix}$ (o cualquier múltiplo escalar).

### Propiedades Importantes

- Una matriz $n \times n$ tiene exactamente $n$ eigenvalores (contando multiplicidades)
- Los eigenvalores de una matriz simétrica son siempre reales
- Los eigenvectores correspondientes a eigenvalores distintos son linealmente independientes
- La **traza** de $A$ (suma de elementos diagonales) es igual a la suma de eigenvalores: $\operatorname{tr}(A) = \sum_{i=1}^n \lambda_i$
- El **determinante** de $A$ es igual al producto de eigenvalores: $\det(A) = \prod_{i=1}^n \lambda_i$

- **Análisis de estabilidad** en sistemas dinámicos
- **Diagonalización** de matrices
- **Análisis de componentes principales (PCA)** en estadística
- **Mecánica cuántica** (valores propios de observables)

## ¿Qué es $\omega^2$ (Omega al Cuadrado)?

### Definición

$\omega^2$ es el **cuadrado de la frecuencia angular** (también llamada frecuencia circular) de un sistema oscilatorio. La frecuencia angular $\omega$ se mide en radianes por segundo (rad/s).

### Relación con la Frecuencia

La frecuencia angular $\omega$ está relacionada con la frecuencia ordinaria $f$ (en Hz) y el período $T$ (en segundos) mediante:

$$\omega = 2\pi f = \frac{2\pi}{T}$$

Por lo tanto:

$$\omega^2 = 4\pi^2 f^2 = \frac{4\pi^2}{T^2}$$

### Conexión con la Ley de Hooke y Eigenvalores

En un sistema masa-resorte simple con constante $k$ y masa $m$, la ecuación de movimiento es:

$$m\frac{d^2x}{dt^2} = -kx$$

La solución es oscilatoria con frecuencia angular:

$$\omega = \sqrt{\frac{k}{m}} \quad \Rightarrow \quad \omega^2 = \frac{k}{m}$$

Aquí $\omega^2$ representa la **razón entre la rigidez y la masa**.

### En Sistemas Multidimensionales

Para sistemas con múltiples grados de libertad (matrices de rigidez $K$ y masa $M$), el problema de eigenvalores es:

$$K\mathbf{v} = \omega^2 M\mathbf{v}$$

o equivalentemente:

$$M^{-1}K\mathbf{v} = \omega^2 \mathbf{v}$$

En este contexto:
- Los **eigenvalores** $\lambda = \omega^2$ son los cuadrados de las frecuencias naturales del sistema
- Los **eigenvectores** $\mathbf{v}$ son los **modos normales** (patrones de vibración)

### Significado Físico de $\omega^2$

| Situación | Interpretación de $\omega^2$ |
|-----------|------------------------------|
| $\omega^2 > 0$ | Sistema estable, oscilaciones armónicas |
| $\omega^2 = 0$ | Sistema en equilibrio neutro, sin restauración |
| $\omega^2 < 0$ | Sistema inestable, crecimiento exponencial (no oscila) |

### Ejemplo Numérico

Si un resorte tiene:
- Constante $k = 100$ N/m
- Masa $m = 4$ kg

Entonces:

$$\omega^2 = \frac{k}{m} = \frac{100}{4} = 25 \text{ rad}^2/\text{s}^2$$

$$\omega = \sqrt{25} = 5 \text{ rad/s}$$

$$f = \frac{\omega}{2\pi} = \frac{5}{2\pi} \approx 0.796 \text{ Hz}$$

$$T = \frac{1}{f} \approx 1.257 \text{ s}$$

### Resumen

$\omega^2$ es un eigenvalor que representa la **frecuencia natural al cuadrado** de un modo de vibración. Es fundamental en el análisis de:
- Vibraciones mecánicas
- Circuitos eléctricos LC
- Óptica (cavidades resonantes)
- Mecánica cuántica (niveles de energía)

## Resolución de la Ecuación de Movimiento Masa-Resorte con Métodos Numéricos

### Ecuación Diferencial

La ecuación de movimiento de un sistema masa-resorte es:

$$m\frac{d^2x}{dt^2} = -kx$$

O equivalentemente:

$$\frac{d^2x}{dt^2} = -\frac{k}{m}x = -\omega^2 x$$

donde $\omega^2 = \frac{k}{m}$.

### Conversión a Sistema de Ecuaciones de Primer Orden

Para aplicar métodos numéricos como Euler o RK4, convertimos la ecuación de segundo orden en un sistema de dos ecuaciones de primer orden.

Definimos:
- $x_1 = x$ (posición)
- $x_2 = v = \frac{dx}{dt}$ (velocidad)

Entonces:

$$\frac{dx_1}{dt} = x_2$$

$$\frac{dx_2}{dt} = -\frac{k}{m}x_1 = -\omega^2 x_1$$

En forma vectorial:

$$\frac{d}{dt}\begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = \begin{pmatrix} x_2 \\ -\omega^2 x_1 \end{pmatrix}$$

O compactamente:

$$\frac{d\mathbf{X}}{dt} = \mathbf{f}(t, \mathbf{X})$$

donde $\mathbf{X} = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}$ y $\mathbf{f}(t, \mathbf{X}) = \begin{pmatrix} x_2 \\ -\omega^2 x_1 \end{pmatrix}$.

---

## Método de Euler

### Fórmula

$$\mathbf{X}_{n+1} = \mathbf{X}_n + h \cdot \mathbf{f}(t_n, \mathbf{X}_n)$$

### Componentes

$$x_{1,n+1} = x_{1,n} + h \cdot x_{2,n}$$

$$x_{2,n+1} = x_{2,n} + h \cdot (-\omega^2 x_{1,n})$$

### Pseudocódigo

```pseudocode
h = paso de tiempo
t = 0
x1 = x0        # posición inicial
x2 = v0        # velocidad inicial
mientras t < t_max:
x1_nuevo = x1 + h * x2
x2_nuevo = x2 + h * (-omega^2 * x1)
x1 = x1_nuevo
x2 = x2_nuevo
t = t + h
guardar resultados
```

### Precisión

- **Orden 1**: Error proporcional a $h$
- Tiende a **inestabilidad** y **deriva de energía** para pasos grandes
- No conserva exactamente la energía mecánica total

---

## Método de Runge-Kutta de 4to Orden (RK4)

### Fórmula General

$$\mathbf{X}_{n+1} = \mathbf{X}_n + \frac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4)$$

donde:

$$k_1 = \mathbf{f}(t_n, \mathbf{X}_n)$$

$$k_2 = \mathbf{f}\left(t_n + \frac{h}{2}, \mathbf{X}_n + \frac{h}{2}k_1\right)$$

$$k_3 = \mathbf{f}\left(t_n + \frac{h}{2}, \mathbf{X}_n + \frac{h}{2}k_2\right)$$

$$k_4 = \mathbf{f}(t_n + h, \mathbf{X}_n + hk_3)$$

### Aplicación al Sistema Masa-Resorte

Para $\mathbf{X} = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}$:

**Paso 1 - $k_1$:**

$$k_{1,1} = x_{2,n}$$

$$k_{1,2} = -\omega^2 x_{1,n}$$

**Paso 2 - $k_2$:**

$$k_{2,1} = x_{2,n} + \frac{h}{2}k_{1,2}$$

$$k_{2,2} = -\omega^2 \left(x_{1,n} + \frac{h}{2}k_{1,1}\right)$$

**Paso 3 - $k_3$:**

$$k_{3,1} = x_{2,n} + \frac{h}{2}k_{2,2}$$

$$k_{3,2} = -\omega^2 \left(x_{1,n} + \frac{h}{2}k_{2,1}\right)$$

**Paso 4 - $k_4$:**

$$k_{4,1} = x_{2,n} + hk_{3,2}$$

$$k_{4,2} = -\omega^2 \left(x_{1,n} + hk_{3,1}\right)$$

**Actualización:**

$$x_{1,n+1} = x_{1,n} + \frac{h}{6}(k_{1,1} + 2k_{2,1} + 2k_{3,1} + k_{4,1})$$

$$x_{2,n+1} = x_{2,n} + \frac{h}{6}(k_{1,2} + 2k_{2,2} + 2k_{3,2} + k_{4,2})$$

### Precisión

- **Orden 4**: Error proporcional a $h^4$
- Mucho más estable que Euler
- Mejor conservación de energía (aunque no perfecta)
- Requiere 4 evaluaciones de la función por paso

| Característica | Euler | RK4 |
|----------------|-------|-----|
| Orden de precisión | 1 | 4 |
| Evaluaciones por paso | 1 | 4 |

## Solución Analítica (para comparación)

La solución exacta es:

$$x(t) = A\cos(\omega t) + B\sin(\omega t)$$

o equivalentemente:

$$x(t) = C\cos(\omega t + \phi)$$

donde las constantes se determinan por condiciones iniciales:

$$A = x_0, \quad B = \frac{v_0}{\omega}$$

Esta solución sirve para verificar la precisión de los métodos numéricos.


## Sistema de Dos Masas con Resortes y Gravedad

### Descripción del Sistema

- $m_1$: masa superior, conectada al techo mediante resorte $k_1$
- $m_2$: masa inferior, conectada a $m_1$ mediante resorte $k_2$
- $y_1, y_2$: posiciones verticales de $m_1$ y $m_2$ (positivo hacia abajo)
- $L_1, L_2$: longitudes naturales de los resortes
- $g$: aceleración gravitacional

### Ecuaciones de Movimiento

**Fuerzas sobre $m_1$:**
- Resorte superior: $-k_1(y_1 - L_1)$ (hacia arriba si estirado)
- Resorte inferior: $k_2(y_2 - y_1 - L_2)$ (hacia abajo si $m_2$ está más abajo)
- Gravedad: $m_1 g$ (hacia abajo)

$$m_1 \ddot{y}_1 = -k_1(y_1 - L_1) + k_2(y_2 - y_1 - L_2) + m_1 g$$

**Fuerzas sobre $m_2$:**
- Resorte inferior: $-k_2(y_2 - y_1 - L_2)$ (hacia arriba si estirado)
- Gravedad: $m_2 g$ (hacia abajo)

$$m_2 \ddot{y}_2 = -k_2(y_2 - y_1 - L_2) + m_2 g$$

### Simplificación: Coordenadas de Desplazamiento desde Equilibrio

Definimos posiciones de equilibrio $y_{1,eq}, y_{2,eq}$ donde las fuerzas netas son cero. Sea $x_1 = y_1 - y_{1,eq}$ y $x_2 = y_2 - y_{2,eq}$.

Las ecuaciones linealizadas alrededor del equilibrio son:

$$m_1 \ddot{x}_1 = -(k_1 + k_2)x_1 + k_2 x_2$$

$$m_2 \ddot{x}_2 = k_2 x_1 - k_2 x_2$$

En forma matricial:

$$\begin{pmatrix} m_1 & 0 \\ 0 & m_2 \end{pmatrix} \begin{pmatrix} \ddot{x}_1 \\ \ddot{x}_2 \end{pmatrix} = \begin{pmatrix} -(k_1+k_2) & k_2 \\ k_2 & -k_2 \end{pmatrix} \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}$$

O equivalentemente:

$$\ddot{\mathbf{x}} = M^{-1}K \mathbf{x}$$

donde:

$$M = \begin{pmatrix} m_1 & 0 \\ 0 & m_2 \end{pmatrix}, \quad K = \begin{pmatrix} -(k_1+k_2) & k_2 \\ k_2 & -k_2 \end{pmatrix}$$

## ¿Solución Analítica o Numérica?

### Solución Analítica: Sí existe

Asumiendo oscilaciones armónicas $x_i(t) = A_i e^{i\omega t}$, obtenemos:

$$-\omega^2 \mathbf{A} = M^{-1}K \mathbf{A}$$

o:

$$(K + \omega^2 M)\mathbf{A} = \mathbf{0}$$

Esto es un **problema de eigenvalores generalizado**:

$$K\mathbf{A} = -\omega^2 M \mathbf{A}$$

Multiplicando por $M^{-1}$:

$$M^{-1}K\mathbf{A} = -\omega^2 \mathbf{A}$$

Los eigenvalores $\lambda = -\omega^2$ dan las frecuencias naturales, y los eigenvectores dan los modos normales.

**La solución analítica general es:**

$$\mathbf{x}(t) = c_1 \mathbf{v}_1 \cos(\omega_1 t + \phi_1) + c_2 \mathbf{v}_2 \cos(\omega_2 t + \phi_2)$$

donde $\omega_1, \omega_2$ son las frecuencias naturales y $\mathbf{v}_1, \mathbf{v}_2$ son los modos normales.

### Cuándo usar Métodos Numéricos

| Situación | Método recomendado |
|-----------|-------------------|
| Sistema lineal (resortes ideales) | Solución analítica con eigenvalores |
| Amplitudes grandes (no linealidad) | RK4 o métodos de integración symplectic |
| Fuerzas externas arbitrarias | RK4 |
- Amortiguamiento viscoso | RK4 |
- Colisiones o restricciones | Métodos de eventos o Euler con proyección |

## Papel de Valores y Vectores Característicos

### 1. Cálculo de Frecuencias Naturales

Resolviendo el problema de eigenvalores:

$$\det(K - \lambda M) = 0$$

Se obtienen dos eigenvalores $\lambda_1, \lambda_2 < 0$, de donde:

$$\omega_1 = \sqrt{-\lambda_1}, \quad \omega_2 = \sqrt{-\lambda_2}$$

### 2. Modos Normales (Eigenvectores)

Cada eigenvector $\mathbf{v}_i$ describe un **modo normal** de vibración:

**Modo 1 ($\omega_1$):** Ambas masas oscilan en fase (se mueven juntas)

$$\mathbf{v}_1 = \begin{pmatrix} v_{11} \\ v_{12} \end{pmatrix}$$

**Modo 2 ($\omega_2$):** Ambas masas oscilan en oposición de fase

$$\mathbf{v}_2 = \begin{pmatrix} v_{21} \\ v_{22} \end{pmatrix}$$

### 3. Desacoplamiento del Sistema

Definiendo la matriz modal $V = [\mathbf{v}_1 \quad \mathbf{v}_2]$ y cambio de coordenadas $\mathbf{x} = V\mathbf{q}$:

$$\ddot{\mathbf{q}} = \Lambda \mathbf{q}$$

donde $\Lambda = \text{diag}(-\omega_1^2, -\omega_2^2)$.

Las coordenadas normales $q_1, q_2$ evolucionan independientemente:

$$\ddot{q}_i = -\omega_i^2 q_i$$

### 4. Condiciones Iniciales

Dadas $\mathbf{x}(0)$ y $\dot{\mathbf{x}}(0)$, las constantes se encuentran proyectando sobre los modos normales:

$$c_i \cos(\phi_i) = \frac{\mathbf{v}_i^T M \mathbf{x}(0)}{\mathbf{v}_i^T M \mathbf{v}_i}$$

## Simulación Numérica con RK4

### Sistema de Primer Orden

Definimos $\mathbf{X} = (x_1, x_2, v_1, v_2)^T$:

$$\frac{d}{dt}\begin{pmatrix} x_1 \\ x_2 \\ v_1 \\ v_2 \end{pmatrix} = \begin{pmatrix} v_1 \\ v_2 \\ -\frac{k_1+k_2}{m_1}x_1 + \frac{k_2}{m_1}x_2 \\ \frac{k_2}{m_2}x_1 - \frac{k_2}{m_2}x_2 \end{pmatrix}$$

### Pseudocódigo RK4
```pseudocode
function f(X):
x1, x2, v1, v2 = X
a1 = -(k1+k2)/m1 * x1 + k2/m1 * x2
a2 = k2/m2 * x1 - k2/m2 * x2
return [v1, v2, a1, a2]
h = paso de tiempo
X = [x1_0, x2_0, v1_0, v2_0]
mientras t < t_max:
k1 = f(X)
k2 = f(X + h/2 * k1)
k3 = f(X + h/2 * k2)
k4 = f(X + h * k3)
X = X + h/6 * (k1 + 2k2 + 2k3 + k4)
t = t + h
guardar X
```

## Resumen: Uso de Eigenvalores en la Simulación

| Aspecto | Rol de Eigenvalores/Vectores |
|---------|------------------------------|
| Frecuencias naturales | $\omega_i = \sqrt{-\lambda_i}$ |
| Modos de vibración | Eigenvectores $\mathbf{v}_i$ |
| Condiciones iniciales | Descomposición en modos normales |
| Validación numérica | Comparar con solución analítica |
| Estabilidad del método | Paso $h$ debe ser $\ll T_{min} = 2\pi/\omega_{max}$ |

Los eigenvalores permiten entender la dinámica fundamental del sistema y verificar la correctitud de la simulación numérica.

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Circle, FancyBboxPatch
import matplotlib.patches as patches
from IPython.display import HTML, display

# Parámetros del sistema
m1 = 1.0  # kg
m2 = 0.8  # kg
k1 = 10.0  # N/m
k2 = 8.0   # N/m
g = 9.81   # m/s^2
L1 = 0.5   # longitud natural resorte 1 (m)
L2 = 0.4   # longitud natural resorte 2 (m)

# Posiciones de equilibrio
# k1*y1_eq = k2*(y2_eq - y1_eq - L2) + m1*g
# k2*(y2_eq - y1_eq - L2) = m2*g

# Resolviendo:
y2_eq_minus_y1_eq = L2 + m2*g/k2
y1_eq = (k2*y2_eq_minus_y1_eq + m1*g)/k1
y2_eq = y1_eq + y2_eq_minus_y1_eq

print(f"Posición de equilibrio m1: {y1_eq:.3f} m")
print(f"Posición de equilibrio m2: {y2_eq:.3f} m")

# Matriz del sistema (coordenadas de desplazamiento desde equilibrio)
# m1*x1'' = -(k1+k2)*x1 + k2*x2
# m2*x2'' = k2*x1 - k2*x2

M = np.array([[m1, 0], [0, m2]])
K = np.array([[-(k1+k2), k2], [k2, -k2]])

# Problema de eigenvalores: K*v = -omega^2 * M*v
# Equivalente a: M^(-1)*K*v = -omega^2*v

Minv_K = np.linalg.inv(M) @ K
eigenvalues, eigenvectors = np.linalg.eig(Minv_K)

# Ordenar por frecuencia (valores propios negativos)
idx = np.argsort(-eigenvalues)  # ordenar de mayor a menor (menos negativo primero)
omega_squared = -eigenvalues[idx]
omega = np.sqrt(omega_squared)
modes = eigenvectors[:, idx]

print(f"\nFrecuencias naturales:")
print(f"ω1 = {omega[0]:.3f} rad/s (f1 = {omega[0]/(2*np.pi):.3f} Hz)")
print(f"ω2 = {omega[1]:.3f} rad/s (f2 = {omega[1]/(2*np.pi):.3f} Hz)")

print(f"\nModos normales:")
print(f"Modo 1: {modes[:, 0]}")
print(f"Modo 2: {modes[:, 1]}")

# Normalizar modos para que el desplazamiento máximo sea 1
for i in range(2):
    modes[:, i] = modes[:, i] / np.max(np.abs(modes[:, i]))

# Condiciones iniciales (perturbación del modo 1)
x0 = np.array([0.1, 0.1])  # desplazamientos iniciales
v0 = np.array([0.0, 0.0])  # velocidades iniciales

# Resolver para amplitudes y fases usando condiciones iniciales
# x(t) = A1*v1*cos(w1*t) + A2*v2*cos(w2*t)
# x(0) = A1*v1 + A2*v2 = V * [A1, A2]^T

V = modes
A = np.linalg.solve(V, x0)

# Velocidades iniciales (suponiendo fases cero por simplicidad, ajustar si es necesario)
# Si v0 != 0, necesitamos términos seno
# Solución general: x(t) = sum(Ai*vi*cos(wi*t) + Bi*vi*sin(wi*t))

B = np.linalg.solve(V, v0 / omega)  # coeficientes para términos seno

# Tiempo de simulación
dt = 0.01
t_max = 10.0
t = np.arange(0, t_max, dt)

# Solución analítica
x1_t = A[0]*modes[0,0]*np.cos(omega[0]*t) + B[0]*modes[0,0]*np.sin(omega[0]*t) + \
       A[1]*modes[0,1]*np.cos(omega[1]*t) + B[1]*modes[0,1]*np.sin(omega[1]*t)

x2_t = A[0]*modes[1,0]*np.cos(omega[0]*t) + B[0]*modes[1,0]*np.sin(omega[0]*t) + \
       A[1]*modes[1,1]*np.cos(omega[1]*t) + B[1]*modes[1,1]*np.sin(omega[1]*t)

# Posiciones absolutas
y1_t = y1_eq + x1_t
y2_t = y2_eq + x2_t

# Configuración de la figura 3D
fig = plt.figure(figsize=(14, 10))

# Vista lateral (2D pero con perspectiva 3D)
ax1 = fig.add_subplot(2, 2, 1)
ax1.set_xlim(-0.5, 0.5)
ax1.set_ylim(0, y2_eq + 0.5)
ax1.set_aspect('equal')
ax1.set_title('Vista Lateral - Animación')
ax1.set_xlabel('x (m)')
ax1.set_ylabel('y (m)')

# Vista 3D
ax2 = fig.add_subplot(2, 2, 2, projection='3d')
ax2.set_xlim(-0.5, 0.5)
ax2.set_ylim(-0.5, 0.5)
ax2.set_zlim(0, y2_eq + 0.5)
ax2.set_title('Vista 3D')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

# Gráficas de posición vs tiempo
ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(t, y1_t, 'b-', label='m1', linewidth=1)
ax3.plot(t, y2_t, 'r-', label='m2', linewidth=1)
ax3.set_xlabel('Tiempo (s)')
ax3.set_ylabel('Posición y (m)')
ax3.set_title('Posiciones vs Tiempo')
ax3.legend()
ax3.grid(True)

# Gráfica de fase
ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(x1_t, x2_t, 'g-', linewidth=1)
ax4.set_xlabel('Desplazamiento m1 (m)')
ax4.set_ylabel('Desplazamiento m2 (m)')
ax4.set_title('Espacio de Fase (x1 vs x2)')
ax4.grid(True)

# Elementos de la animación (vista lateral)
ceiling, = ax1.plot([-0.3, 0.3], [0, 0], 'k-', linewidth=3)
spring1, = ax1.plot([], [], 'b-', linewidth=2)
mass1 = Circle((0, y1_eq), 0.08, color='blue', fill=True)
ax1.add_patch(mass1)
spring2, = ax1.plot([], [], 'r-', linewidth=2)
mass2 = Circle((0, y2_eq), 0.06, color='red', fill=True)
ax1.add_patch(mass2)
trail1, = ax1.plot([], [], 'b--', alpha=0.3, linewidth=1)
trail2, = ax1.plot([], [], 'r--', alpha=0.3, linewidth=1)

# Elementos 3D
spring1_3d, = ax2.plot([], [], [], 'b-', linewidth=2)
spring2_3d, = ax2.plot([], [], [], 'r-', linewidth=2)
mass1_3d = ax2.scatter([], [], [], c='blue', s=200, marker='o')
mass2_3d = ax2.scatter([], [], [], c='red', s=150, marker='o')

# Historial para rastros
trail_length = 50
y1_history = []
y2_history = []

def draw_spring(ax, y_top, y_bottom, x_pos=0, n_coils=10, width=0.1):
    """Dibuja un resorte como línea zigzagueante"""
    y = np.linspace(y_top, y_bottom, n_coils*2 + 1)
    x = np.zeros_like(y)
    x[1:-1:2] = width
    x[2:-1:2] = -width
    x += x_pos
    return x, y

def draw_spring_3d(ax, y_top, y_bottom, x_pos=0, y_pos=0, n_coils=10, radius=0.1):
    """Dibuja un resorte en 3D (espiral)"""
    t = np.linspace(0, 2*np.pi*n_coils, 100)
    y = np.linspace(y_top, y_bottom, 100)
    x = x_pos + radius * np.cos(t)
    z = y_pos + radius * np.sin(t)
    return x, z, y

def init():
    spring1.set_data([], [])
    spring2.set_data([], [])
    mass1.center = (0, y1_eq)
    mass2.center = (0, y2_eq)
    trail1.set_data([], [])
    trail2.set_data([], [])
    
    spring1_3d.set_data([], [])
    spring1_3d.set_3d_properties([])
    spring2_3d.set_data([], [])
    spring2_3d.set_3d_properties([])
    mass1_3d._offsets3d = ([0], [0], [y1_eq])
    mass2_3d._offsets3d = ([0], [0], [y2_eq])
    
    return spring1, spring2, mass1, mass2, trail1, trail2, spring1_3d, spring2_3d, mass1_3d, mass2_3d

def update(frame):
    y1 = y1_t[frame]
    y2 = y2_t[frame]
    
    # Actualizar vista lateral
    x_spring1, y_spring1 = draw_spring(ax1, 0, y1, n_coils=8)
    spring1.set_data(x_spring1, y_spring1)
    
    x_spring2, y_spring2 = draw_spring(ax1, y1, y2, n_coils=6)
    spring2.set_data(x_spring2, y_spring2)
    
    mass1.center = (0, y1)
    mass2.center = (0, y2)
    
    # Actualizar rastros
    y1_history.append(y1)
    y2_history.append(y2)
    if len(y1_history) > trail_length:
        y1_history.pop(0)
        y2_history.pop(0)
    
    trail1.set_data([0]*len(y1_history), y1_history)
    trail2.set_data([0]*len(y2_history), y2_history)
    
    # Actualizar vista 3D
    x_s1, z_s1, y_s1 = draw_spring_3d(ax2, 0, y1, n_coils=8)
    spring1_3d.set_data(x_s1, z_s1)
    spring1_3d.set_3d_properties(y_s1)
    
    x_s2, z_s2, y_s2 = draw_spring_3d(ax2, y1, y2, n_coils=6)
    spring2_3d.set_data(x_s2, z_s2)
    spring2_3d.set_3d_properties(y_s2)
    
    mass1_3d._offsets3d = ([0], [0], [y1])
    mass2_3d._offsets3d = ([0], [0], [y2])
    
    # Actualizar línea de tiempo en gráfica de posiciones
    ax3.axvline(x=t[frame], color='gray', linestyle='--', alpha=0.5)
    
    return spring1, spring2, mass1, mass2, trail1, trail2, spring1_3d, spring2_3d, mass1_3d, mass2_3d

# Crear animación
anim = FuncAnimation(fig, update, frames=len(t), init_func=init,
                     interval=dt*1000, blit=False, repeat=True)

plt.tight_layout()
plt.show()

# Mostrar animación en JupyterLite
display(HTML(anim.to_jshtml()))

# Guardar información del sistema
print(f"\n--- Parámetros del Sistema ---")
print(f"m1 = {m1} kg, m2 = {m2} kg")
print(f"k1 = {k1} N/m, k2 = {k2} N/m")
print(f"Periodo T1 = {2*np.pi/omega[0]:.3f} s")
print(f"Periodo T2 = {2*np.pi/omega[1]:.3f} s")
````
