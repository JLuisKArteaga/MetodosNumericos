```p5js
// Variables en window para que persistan entre celdas
window.pos = null;
window.vel = null;
window.acc = null;
window.objetivo = null;

function setup() {
  createCanvas(600, 400);
  
  // Inicializar en window
  window.pos = createVector(50, 200);
  window.vel = createVector(2, 0);
  window.acc = createVector(0, 0);
  window.objetivo = createVector(500, 200);
}
```

```p5js
function draw() {
  background(240, 240, 240, 50);
  
  // Acceder a variables globales
  var p = window.pos;
  var v = window.vel;
  var o = window.objetivo;
  
  // Verificar que existen
  if (!p || !v || !o) {
    text("Error: variables no inicializadas", 50, 50);
    return;
  }
  
  // Calcular dirección
  var dirX = o.x - p.x;
  var dirY = o.y - p.y;
  var dist = Math.sqrt(dirX * dirX + dirY * dirY);
  
  // Normalizar y escalar
  if (dist > 0) {
    dirX = (dirX / dist) * 0.2;
    dirY = (dirY / dist) * 0.2;
  }
  
  // Actualizar física
  v.x += dirX;
  v.y += dirY;
  v.x *= 0.98;
  v.y *= 0.98;
  p.x += v.x;
  p.y += v.y;
  
  // Dibujar
  fill(0, 255, 0);
  noStroke();
  ellipse(o.x, o.y, 20, 20);
  
  fill(255, 100, 100);
  ellipse(p.x, p.y, 20, 20);
  
  stroke(200);
  line(p.x, p.y, o.x, o.y);
  
  stroke(255, 0, 0);
  strokeWeight(2);
  line(p.x, p.y, p.x + v.x * 10, p.y + v.y * 10);
  
  stroke(0, 0, 255);
  line(p.x, p.y, p.x + dirX * 50, p.y + dirY * 50);
  
  fill(0);
  noStroke();
  textSize(12);
  text("Pos: (" + p.x.toFixed(1) + "," + p.y.toFixed(1) + ")", 20, 30);
  text("Dist: " + dist.toFixed(1), 20, 50);
  
  if (dist < 10) {
    window.pos = createVector(50, random(100, 300));
    window.vel = createVector(2, 0);
  }
}
```
