```p5js
// PELOTA REBOTANDO CON GRAVEDAD Y RESTITUCIÓN

window.pelota = null;
window.gravedad = null;
window.sueloY = null;
window.restitucion = null;

function setup() {
  createCanvas(600, 400);
  
  // Pelota: posición, velocidad, radio
  window.pelota = {
    pos: createVector(100, 50),
    vel: createVector(5, 0),
    r: 20
  };
  
  // Gravedad (aceleración hacia abajo)
  window.gravedad = createVector(0, 0.5);
  
  // Posición del suelo
  window.sueloY = 350;
  
  // Coeficiente de restitución
  window.restitucion = 0.8;
  
  console.log("Setup completado");
}

"Listo - ejecutar celda 2"
```

```p5js
function draw() {
  // Verificación de seguridad
  if (!window.pelota || !window.gravedad) {
    background(200);
    fill(0);
    textSize(20);
    textAlign(CENTER, CENTER);
    text("Ejecutar Celda 1 primero", width/2, height/2);
    return;
  }
  
  var b = window.pelota;
  var g = window.gravedad;
  var suelo = window.sueloY;
  var e = window.restitucion;
  
  background(240);
  
  // Aplicar gravedad
  b.vel.add(g);
  
  // Actualizar posición
  b.pos.add(b.vel);
  
  // Colisión con suelo
  if (b.pos.y > suelo - b.r) {
    b.pos.y = suelo - b.r;
    b.vel.y *= -e;
    b.vel.x *= 0.99;
  }
  
  // Rebote en paredes
  if (b.pos.x < b.r) {
    b.pos.x = b.r;
    b.vel.x *= -e;
  }
  if (b.pos.x > width - b.r) {
    b.pos.x = width - b.r;
    b.vel.x *= -e;
  }
  
  // Dibujar suelo
  stroke(100);
  strokeWeight(2);
  line(0, suelo, width, suelo);
  
  // Dibujar pelota
  fill(255, 100, 100);
  noStroke();
  ellipse(b.pos.x, b.pos.y, b.r * 2);
  
  // Vector velocidad
  stroke(255, 0, 0);
  strokeWeight(2);
  line(b.pos.x, b.pos.y, b.pos.x + b.vel.x * 3, b.pos.y + b.vel.y * 3);
  
  // Información
  fill(0);
  noStroke();
  textAlign(LEFT);
  textSize(12);
  text("GRAVEDAD Y RESTITUCIÓN", 20, 30);
  text("Pos: (" + b.pos.x.toFixed(1) + ", " + b.pos.y.toFixed(1) + ")", 20, 50);
  text("Vel: (" + b.vel.x.toFixed(2) + ", " + b.vel.y.toFixed(2) + ")", 20, 70);
  text("Restitución (e): " + e, 20, 90);
  
  // Reiniciar si se detiene
  if (Math.abs(b.vel.y) < 0.1 && b.pos.y > suelo - b.r - 1) {
    b.pos = createVector(random(50, width - 50), 50);
    b.vel = createVector(random(-5, 5), 0);
  }
}
```
