# Sistemas de ecuaciones lineales

## 4-1 Introducción

La necesidad de tener métodos eficientes para resolver sistemas de ecuaciones algebraicas lineales, es más importante aún que la solución de las ecuaciones no lineales. Esto es debido a que en muchos casos los problemas de matemáticas aplicadas se reducen a un conjunto de ecuaciones que constituyen un sistema lineal. Por este motivo puede decirse con razón que la principal tarea del análisis numérico es la solución de los sistemas de ecuaciones lineales.

Se ha desarrollado una extraordinaria colección de algoritmos para llevar a cabo la solución de los sistemas lineales, lo que indica que es engañoso el aparente carácter elemental del problema, además de que hay muchas fallas en los métodos existentes. En forma semejante a las ecuaciones no lineales, cuyos métodos de solución se estudiaron en el capítulo anterior, los sistemas lineales pueden resolverse por medio de métodos exactos y de aproximaciones sucesivas. Antes de establecer estos métodos, se recordarán algunos conceptos fundamentales sobre los sistemas lineales.

1. **Ecuación algebraica lineal.** Una *ecuación algebraica lineal* es aquella en donde en cada término de la ecuación aparece únicamente una variable o incógnita elevada a la primera potencia. Por ejemplo,

$$a_{11}x_{1} + a_{12}x_{2} + a_{13}x_{3} + \ldots + a_{1n}x_{n} = b_{1} \quad (4\text{-}1)$$

es una ecuación algebraica lineal en las variables $x_{1}, x_{2}, x_{3}, \ldots, x_{n}$ . Se admite que los coeficientes $a_{11}, a_{12}, a_{13}, \ldots, a_{1n}$ y el término independiente $b_{1}$, de la ecuación, son constantes reales.

2. **Sistema de ecuaciones.** Un *sistema de ecuaciones* es un conjunto de ecuaciones que deben resolverse simultáneamente. En lo sucesivo se considerarán únicamente sistemas de ecuaciones algebraicas lineales, o sea conjuntos de ecuaciones de la forma

$$
\begin{cases}
a_{11}x_{1} + a_{12}x_{2} + a_{13}x_{3} + \ldots + a_{1n}x_{n} = b_{1} \\
a_{21}x_{1} + a_{22}x_{2} + a_{23}x_{3} + \ldots + a_{2n}x_{n} = b_{2} \\
a_{31}x_{1} + a_{32}x_{2} + a_{33}x_{3} + \ldots + a_{3n}x_{n} = b_{3} \\
\quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \\
\quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \quad \cdot \\
a_{n1}x_{1} + a_{n2}x_{2} + a_{n3}x_{3} + \ldots + a_{nn}x_{n} = b_{n}
\end{cases} \quad (4\text{-}2)
$$

Aplicando la definición de producto entre matrices, este sistema de $n$ ecuaciones algebraicas lineales con $n$ incógnitas puede escribirse en la forma matricial

$$
\begin{bmatrix} 
a_{11} & a_{12} & a_{13} & \ldots & a_{1n} \\
a_{21} & a_{22} & a_{23} & \ldots & a_{2n} \\
a_{31} & a_{32} & a_{33} & \ldots & a_{3n} \\
\cdot & \cdot & \cdot & \cdot & \cdot \\
\cdot & \cdot & \cdot & \cdot & \cdot \\
a_{n1} & a_{n2} & a_{n3} & \ldots & a_{nn}
\end{bmatrix}
\begin{bmatrix} 
x_{1} \\
x_{2} \\
x_{3} \\
\ldots \\
\ldots \\
x_{n}
\end{bmatrix}
=
\begin{bmatrix} 
b_{1} \\
b_{2} \\
b_{3} \\
\ldots \\
\ldots \\
b_{n}
\end{bmatrix} \quad (4\text{-}3)
$$

en donde $i = 1, 2, 3, \ldots, n$; $j = 1, 2, 3, \ldots, n$. El sistema de ecuaciones (4-3) puede escribirse simbólicamente como

$$\mathbf{A}\mathbf{x} = \mathbf{b} \quad (4\text{-}4)$$

en donde $\mathbf{A}$ se llama la *matriz del sistema*. La matriz formada por $\mathbf{A}$, a la que se le ha agregado el vector de términos independientes como última columna, se le llama la *matriz ampliada* del sistema, que se representa con $(\mathbf{A}, \mathbf{b})$.

3. **Solución de un sistema de ecuaciones.** La solución de un sistema de ecuaciones es un conjunto de valores de las incógnitas que verifican simultáneamente a todas y cada una de las ecuaciones del sistema.
De acuerdo con su solución, un sistema puede ser: compatible, si admite solución; o incompatible, si no admite solución.
Un sistema compatible puede ser: determinado, si la solución es única; o indeterminado, si la solución no es única. En este caso se demuestra que existe una infinidad de soluciones.
4. Teoremas sobre rangos. El rango de una matriz es el orden del determinante no nulo de mayor orden que puede obtenerse de esa matriz. El rango de la matriz A  se representa con la notación r(A)  y el de la matriz ampliada con r(A,b) .
En álgebra se demuestra que
a) Para cualquier sistema r(A)≤r(A,b) .
b) Si r(A)<r(A,b)  el sistema es incompatible.
c) Si r(A)=r(A,b)  el sistema de ecuaciones es compatible. En este caso, si además r(A)=n , el sistema es determinado; e indeterminado si r(A)<n , siendo n  el número de variables en el sistema.
## Ejemplo 4-1 Para el sistema de ecuaciones

$$\begin{cases} 3x_1 + 2x_2 = 5 \\ 2x_1 + 4x_2 = 6 \end{cases}$$


el sistema de valores x $x_1=1, x_2=1$, de las incógnitas, lo verifican idénticamente, luego es una solución de él, y puede afirmarse que el sistema es compatible. Para este problema puede demostrarse que $r(\mathbf{A}) = r(\mathbf{A}, \mathbf{b}) = 2$ , por lo que el sistema es compatible, como se indicó, y además determinado, o sea que la solución mencionada es la única que admite.
En el caso de 
$$
\begin{cases}
3x_1 + 2x_2 = 5 \\[6pt]
3x_1 + 2x_2 = 10
\end{cases} \quad (4\text{-}6)
$$

no existe ningún sistema de valores para $x_1$  y $x_2$ que hagan que 5 sea igual a 10 por lo que el sistema es incompatible. Para este sistema puede demostrarse $r(A)=1$ y que $r(A,b)=2$.
Para el sistema
$$
\begin{cases}
3x_1 + 2x_2 = 5 \\[6pt]
6x_1 + 4x_2 = 10
\end{cases} \quad (4\text{-}7)
$$

se puede observar que $x_1 = 1, x_2 = 1$ es solución del sistema, por lo que es compatible. También, $x_1 = 0$, $x_2 = 5/2$ es solución del sistema, por lo que es indeterminado. Fácilmente puede verse que cualquier sistema de valores de $x_1$ y $x_2 = (5-3x_1)/2$ es solución del sistema.
Para éste se demuestra que $r(\mathbf{A}) = r(\mathbf{A}, \mathbf{b}) = 1 < 2$; es decir, el sistema es compatible indeterminado como se indicó.

La solución numérica de un sistema de ecuaciones siempre debe sustituirse en las ecuaciones originales, para garantizar la validez de la solución obtenida. <br>

**5. Combinación lineal de las ecuaciones de un sistema.** Una *combinación lineal* de las ecuaciones de un sistema es una ecuación obtenida al sumar miembro a miembro algunas de las ecuaciones del sistema, multiplicando previamente cada una de ellas por constantes.

Por ejemplo, si en el sistema de ecuaciones (4-5) se multiplica la primera ecuación por 3 y la segunda por 2, y se suman miembro a miembro las ecuaciones, se obtiene

$$3(3x_1 + 2x_2) + 2(2x_1 + 4x_2) = 3(5) + 2(6)$$

de donde

$$13x_1 + 14x_2 = 27 \quad (4\text{-}8)$$

Es una combinación lineal de las ecuaciones del sistema considerado. <br>
**6. Ecuación consecuencia de un sistema.** Una *ecuación consecuencia de un sistema* es aquella ecuación que se verifica para el mismo sistema de valores de las incógnitas que constituye la solución del sistema.

Por ejemplo, para el sistema de ecuaciones $(4-5)$ se sabe que $x_1 = 1$ y $x_2 = 1$ es su solución, por la que cualquier otra ecuación que se verifique para estos valores será una consecuencia del sistema. Así $6x_1 - 5x_2 = 1$ es una consecuencia del sistema, y también $13x_1 + 14x_2 = 27$ es consecuencia del sistema.

Es fácil observar que *toda combinación lineal de las ecuaciones de un sistema es una consecuencia del mismo*.

**7. Sistemas equivalentes.** Dos sistemas son *equivalentes* si admiten ambos como solución al mismo sistema de valores de las incógnitas.

Por ejemplo, los sistemas

$$
\begin{cases}
3x_1 + 2x_2 = 5 \\[4pt]
2x_1 + 4x_2 = 6
\end{cases}
\quad \text{y} \quad
\begin{cases}
3x_1 + 2x_2 = 5 \\[4pt]
13x_1 + 14x_2 = 27
\end{cases} \quad (4-9)
$$

son equivalentes, ya que el sistema de valores $x_1 = 1$ y $x_2 = 1$ es la única solución de los dos sistemas. Obsérvese que el primer sistema es el presentado en las ecuaciones $(4-5)$ y el segundo ha sido obtenido de éste, sustituyendo su segunda ecuación por una combinación lineal de las ecuaciones del sistema que, se insiste, es una consecuencia del sistema.

**8. Teorema fundamental de equivalencia.** Este teorema se expresa así: *Si en un sistema de ecuaciones se sustituye una de ellas por una combinación lineal de las ecuaciones del sistema, se obtiene un nuevo sistema que es equivalente al anterior*.

Este teorema resulta evidente al considerar que la combinación lineal de las ecuaciones del sistema que se introduce es consecuencia del sistema, por lo que seguirá admitiendo la misma solución.

Por ejemplo, si en el sistema

$$
\begin{cases}
3x_1 + 2x_2 = 5 \\[4pt]
2x_1 + 4x_2 = 6
\end{cases} \quad (4\text{-}10)
$$

se sustituye la segunda ecuación por una combinación lineal de las ecuaciones del sistema, formada al sumar miembro a miembro a la segunda ecuación la primera multiplicada por (−2) .
Se obtiene.

$$
\begin{cases}
3x_1 + 2x_2 = 5 \\[4pt]
-4x_1       = -4
\end{cases} \quad (4\text{-}11)
$$

que es equivalente al sistema original. Obsérvese que el sistema $(4\text{-}11)$ también se verifica para $x_1 = 1, x_2 = 1$, como debía ser. Además, se puede notar que en el nuevo sistema no aparece la variable $x_2$ en la segunda ecuación (se dice que se ha *eliminado* de la segunda ecuación), con lo que prácticamente está resuelto el sistema.

**9. Método de eliminación completa.** El método consiste en aplicar el Teorema Fundamental de equivalencia de manera de ir sustituyendo ecuaciones del sistema por combinaciones lineales del mismo en las que se hayan eliminado incógnitas, hasta que, en el caso más general, quede una sola incógnita en cada ecuación. Con esto podrá despejarse el valor de una incógnita de cada ecuación y con ello obtener la solución del sistema.

## Ejemplo 4-2

Resolver el sistema

$$
\begin{cases}
3x_1 + 2x_2 = 5 \\
2x_1 + 4x_2 = 6
\end{cases}
\quad (4.12)
$$

Se va a anotar abajo del sistema, notacionalmente, la operación que se va a realizar sobre éste para obtener su equivalente. Por ejemplo, $2E + 1E(-2)$, significará que se va a sumar a la segunda ecuación la primera multiplicada por $(-2)$. Se obtiene

**Paso 1:** Sistema original

$$
\begin{cases}
3x_1 + 2x_2 = 5 \\
2x_1 + 4x_2 = 6
\end{cases}
$$

Operación aplicada: $2E + 1E(-2)$

**Paso 2:** Eliminación en la segunda ecuación

$$
\begin{cases}
3x_1 + 2x_2 = 4 \\
-4x_1 = -4
\end{cases}
$$

Operación aplicada: $2E(-1/4)$

**Paso 3:** Normalización y sustitución hacia atrás

$$
\begin{cases}
3x_1 + 2x_2 = 4 \\
x_1 = 1
\end{cases}
$$

Operación aplicada: $1E + 2E(-3)$

**Paso 4:** Sustitución y normalización final

$$
\begin{cases}
2x_2 = 2 \\
x_1 = 1
\end{cases}
\quad \xrightarrow{1E(1/2)} \quad
\begin{cases}
x_2 = 1 \\
x_1 = 1
\end{cases}
$$

de donde se obtiene que la solución única del sistema es $x_1 = 1$, $x_2 = 1$

## Ejemplo 4-3

Resolver el sistema

$$
\begin{cases}
x_1 - x_2 + x_3 = -4 \\
5x_1 - 4x_2 + 3x_3 = -12 \\
2x_1 + x_2 + x_3 = 11
\end{cases}
\quad (4.13)
$$

Aplicando el método de eliminación completa, se tiene

**Etapa 1:** Eliminación de $x_1$ en ecuaciones 2 y 3  
*(Operaciones: $2E + 1E(-5)$ y $3E + 1E(-2)$)*

$$
\begin{cases}
x_1 - x_2 + x_3 = -4 \\
5x_1 - 4x_2 + 3x_3 = -12 \\
2x_1 + x_2 + x_3 = 11
\end{cases}
\longrightarrow
\begin{cases}
x_1 - x_2 + x_3 = -4 \\
x_2 - 2x_3 = 8 \\
3x_2 - x_3 = 19
\end{cases}
$$

**Etapa 2:** Eliminación de $x_2$ en ecuación 3  
*(Operación: $3E + 2E(-3)$)*

$$
\begin{cases}
x_1 - x_2 + x_3 = -4 \\
x_2 - 2x_3 = 8 \\
3x_2 - x_3 = 19
\end{cases}
\longrightarrow
\begin{cases}
x_1 - x_3 = 4 \\
x_2 - 2x_3 = 8 \\
5x_3 = -5
\end{cases}
$$

**Etapa 3:** Normalización y sustitución retroactiva  
*(Operaciones: $3E(1/5) = 3E'$, luego $1E + 3E'$ y $2E + 3E'(2)$)*

$$
\begin{cases}
x_1 - x_3 = 4 \\
x_2 - 2x_3 = 8 \\
5x_3 = -5
\end{cases}
\longrightarrow
\begin{cases}
x_1 = 3 \\
x_2 = 6 \\
x_3 = -1
\end{cases}
$$
$$
3E(1/5) = 3E'
$$

$$
1E + 3E'
$$

$$
2E + 3E'(2)
$$

Debe aclararse que, para obtener el último sistema equivalente, primero se dividió la tercera ecuación entre cinco (representada notacionalmente la operación con $3E(1/5)=3E'$ y después se sumó a la primera ecuación esta tercera ecuación corregida, $1E + 3E'$, y a la segunda ecuación se le sumó también la tercera ecuación corregida multiplicada por dos, $2E + 3E'(2)$. El último sistema determina la solución única del sistema, que es

$$x_1 = 3, \quad x_2 = 6, \quad x_3 = -1.$$

**4-2 Método de Gauss-Jordan**

El método de Gauss-Jordan puede decirse que es un método *exacto* para resolver sistemas lineales. Básicamente consiste en sistematizar el Teorema Fundamental de equivalencia.

Aunque como se indicó, el método es exacto, dista mucho de proporcionar una buena aproximación a la solución del sistema en algunos casos, como se verá más adelante.

Observando el ejemplo 4-3, resuelto con el método de eliminación completa, puede verse que el sistema original y sus equivalentes están perfectamente ordenados; es decir, que en todos aparecen, en una primera columna, coeficientes de $x_1$ y la propia incógnita; en la segunda coeficientes de $x_2$ y la incógnita; en la tercera coeficientes de $x_3$ y esta incógnita, y en la última los términos independientes precedidos del signo igual. Conociendo esto, es posible ahorrarse el trabajo de escribir repetidamente $x_1$, $x_2$, $x_3$ y signos de igual en las ecuaciones, considerando únicamente los coeficientes de las incógnitas y los términos independientes. Al hacerlo, se forma un arreglo de números, o matriz, en donde en la primera columna aparecerán coeficientes de $x_1$, en la segunda de $x_2$, en la tercera de $x_3$ y en la última términos independientes, o sea la matriz ampliada del sistema. Así, el sistema original del ejemplo 4-3 puede escribirse como:

$$
\left[\begin{array}{ccc|c}
1 & -1 & 1 & -4 \\
5 & -4 & 3 & -12 \\
2 & 1 & 1 & 11
\end{array}\right]
$$
en donde se han separado con una línea vertical los coeficientes de las incógnitas en el sistema y los términos independientes. Cualquier renglón de esta matriz ampliada puede interpretarse muy fácilmente como una ecuación del sistema correspondiente. Por ejemplo, el segundo renglón significa que cinco veces $x_1$ menos cuatro veces $x_2$ más tres veces $x_3$ es igual a $-12$, o sea la segunda ecuación.

La solución de un sistema lineal puede obtenerse más fácilmente si se representa en esta forma matricial. La eliminación de incógnitas por medio de la sustitución de ecuaciones por combinaciones lineales, corresponde a tratar de convertir en ceros los elementos de la matriz ampliada, representativos de los coeficientes del sistema. Observando la forma en que se obtuvieron los sistemas equivalente en el ejemplo 4.3, puede aceptarse que las siguientes tres operaciones sobre una matriz ampliada producen otra correspondiente a un sistema equivalente al de la anterior:

1. Intercambiar dos renglones (ya que corresponde a reordenar las ecuaciones del sistema)

2. Multiplicar todos los elementos de un renglón por una misma constante no nula (ya que corresponde a multiplicar ambos miembros de una ecuación por una constante)

3. Sumar a los elementos de un renglón los correspondientes elementos de otro multiplicados por una constante (ya que corresponde a la sustitución de una ecuación por una combinación lineal de las ecuaciones del sistema).

Las operaciones anteriores reciben el nombre de *operaciones elementales sobre los renglones de una matriz*. Para resolver un sistema lineal, se aplican éstas a la matriz ampliada del sistema, de manera de convertir en ceros el mayor número de los elementos correspondientes a los coeficientes; cuando no se pueda anular ningún elemento más, se procede a interpretar los renglones de la última matriz obtenida, con lo que se obtendrá la solución del sistema. Este método recibe el nombre de eliminación completa de Gauss-Jordan.

La aplicación de este método resulta más efectiva si se toma un elemento como *pivote* y se eliminan todos los elementos de su columna, haciendo previamente que el pivote se convierta en uno, en forma sistemática.

**Ejemplo 4-4**

Resolver con el método de eliminación completa de Gauss-Jordan el sistema del ejemplo $(4\text{-}3)$. La matriz ampliada es:

$$
\left[\begin{array}{ccc|c}
\enclose{circle}{1} & -1 & 1 & -4 \\
5 & -4 & 3 & -12 \\
2 & 1 & 1 & 11
\end{array}\right]
$$

Tomando como pivote el elemento circulado, el cual ya vale uno, y sumando a los elementos del segundo renglón los del primero multiplicados por $-5$, notacionalmente $2R + 1R(-5)$, y sumando a los elementos del tercer renglón también los del primero por $(-2)$, notacionalmente $3R + 1R(-2)$, se obtiene:

$$
\left[\begin{array}{ccc|c}
1 & -1 & 1 & -4 \\
0 & \enclose{circle}{1} & -2 & 8 \\
0 & 3 & -1 & 19
\end{array}\right]
$$

Aplicando ahora $1R + 2R(1)$ y $3R + 2R(-3)$

$$
\left[\begin{array}{ccc|c}
1 & 0 & -1 & 4 \\
0 & 1 & -2 & 8 \\
0 & 0 & \enclose{circle}{5} & -5
\end{array}\right]
$$

Finalmente haciendo $3R(1/5) = 3R'$ para convertir en uno el pivote, y posteriormente $1R + 3R'(1)$ y $2R + 3R'(2)$, donde $3R'$ representa el tercer renglón después de multiplicarlo por $1/5$.

$$
\left[\begin{array}{ccc|c}
1 & 0 & 0 & 3 \\
0 & 1 & 0 & 6 \\
0 & 0 & 1 & -1
\end{array}\right]
$$

Como no es posible eliminar ningún elemento más, se procede a interpretar los renglones de la matriz, obteniéndose $x_1 = 3$, $x_2 = 6$, $x_3 = -1$, que es la solución del sistema.



```python
import numpy as np

def gauss_jordan_augmentada(A, b, verbose=True):
    """
    Resuelve Ax = b por Gauss-Jordan usando matriz aumentada [A|b].
    """
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float).reshape(-1, 1)
    n = len(b)
    
    # Crear matriz aumentada [A|b]
    M = np.hstack([A, b])
    
    if verbose:
        print("Matriz aumentada inicial [A|b]:")
        imprimir_matriz(M)
        print("-" * 50)
    
    for i in range(n):
        # Pivoteo parcial
        max_idx = np.argmax(np.abs(M[i:, i])) + i
        
        if M[max_idx, i] == 0:
            raise ValueError("Sistema sin solución única")
        
        if max_idx != i:
            M[[i, max_idx]] = M[[max_idx, i]]
            if verbose:
                print(f"\nIntercambio R{i+1} ↔ R{max_idx+1}:")
                imprimir_matriz(M)
        
        # Normalizar fila pivote
        pivote = M[i, i]
        M[i, :] = M[i, :] / pivote
        
        if verbose:
            print(f"\nR{i+1} ÷ {pivote:.2f} (pivote = 1):")
            imprimir_matriz(M)
        
        # Eliminación en todas las filas (arriba y abajo)
        for j in range(n):
            if j != i and M[j, i] != 0:
                factor = M[j, i]
                M[j, :] = M[j, :] - factor * M[i, :]
                
                if verbose:
                    print(f"\nR{j+1} - ({factor:.2f})×R{i+1}:")
                    imprimir_matriz(M)
    
    x = M[:, -1]
    
    if verbose:
        print("=" * 50)
        print("FORMA ESCALONADA REDUCIDA:")
        imprimir_matriz(M)
        print("\nSOLUCIÓN:")
        for i, val in enumerate(x):
            print(f"x_{i+1} = {val:.4f}")
    
    return x

def imprimir_matriz(M, precision=2):
    """Imprime matriz aumentada con línea vertical."""
    n = M.shape[0]
    A_part = M[:, :-1]
    b_part = M[:, -1]
    
    for i in range(n):
        fila_str = "["
        for j in range(A_part.shape[1]):
            fila_str += f"{A_part[i,j]:8.{precision}f} "
        fila_str = fila_str.rstrip()
        fila_str += f" | {b_part[i]:8.{precision}f}]"
        print(fila_str)

# EJEMPLO DE USO
A = [
    [1, -1, 1],
    [5, -4, 3],
    [2, 1, 1]
]

b = [-4, -12, 11]

print("=" * 60)
print("GAUSS-JORDAN CON MATRIZ AUMENTADA")
print("=" * 60)

solucion = gauss_jordan_augmentada(A, b)

# Verificación
print("\n" + "=" * 60)
print("VERIFICACIÓN A·x = b")
print("=" * 60)
A_np = np.array(A)
resultado = A_np @ solucion
print(f"A·x = {resultado}")
print(f"b    = {b}")
print(f"Error: {np.linalg.norm(resultado - b):.2e}")
```
