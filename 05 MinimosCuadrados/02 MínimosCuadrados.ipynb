## ¿Qué es el Método de Mínimos Cuadrados?

El **método de mínimos cuadrados** es una técnica matemática de optimización que busca encontrar la mejor aproximación de un conjunto de datos mediante una función, minimizando la suma de los cuadrados de las diferencias (residuos) entre los valores observados y los valores predichos por el modelo.

### Concepto Central
Dado un conjunto de puntos $(x_i, y_i)$ con $i = 1, 2, \ldots, n$, buscamos una función $f(x)$ tal que:

$$S = \sum_{i=1}^{n} [y_i - f(x_i)]^2$$

sea **mínima**. Esta suma $S$ representa el **error cuadrático total**.

## (a) Mínimos Cuadrados con Tendencia Lineal

### Modelo Lineal
La forma más simple es ajustar una **recta** a los datos:

$$y = mx + b$$

donde:
- $m$ = pendiente de la recta
- $b$ = intercepto (ordenada al origen)

### Relación con las Derivadas

Para encontrar los valores óptimos de $m$ y $b$ que minimizan el error, aplicamos **cálculo diferencial**:

$$S(m,b) = \sum_{i=1}^{n} (y_i - mx_i - b)^2$$

#### Condición de Mínimo (Derivadas Parciales)

El mínimo se encuentra cuando las derivadas parciales respecto a $m$ y $b$ son **cero**:

$$\frac{\partial S}{\partial m} = -2\sum_{i=1}^{n} x_i(y_i - mx_i - b) = 0$$

$$\frac{\partial S}{\partial b} = -2\sum_{i=1}^{n} (y_i - mx_i - b) = 0$$

### Sistema de Ecuaciones Normales

De las derivadas obtenemos el **sistema de ecuaciones normales**:

$$m\sum_{i=1}^{n} x_i^2 + b\sum_{i=1}^{n} x_i = \sum_{i=1}^{n} x_i y_i$$

$$m\sum_{i=1}^{n} x_i + nb = \sum_{i=1}^{n} y_i$$

### Relación con la Pendiente

La pendiente $m$ representa la **tasa de cambio promedio** de $y$ respecto a $x$. Matemáticamente:

$$m = \frac{n\sum x_i y_i - \sum x_i \sum y_i}{n\sum x_i^2 - (\sum x_i)^2} = \frac{\text{Covarianza}(x,y)}{\text{Varianza}(x)}$$

&gt; **Interpretación geométrica**: La pendiente indica cuánto cambia $y$ por cada unidad de cambio en $x$, calculada de forma que la suma de distancias verticales al cuadrado sea mínima.

## (b) Regresión Lineal

### Definición Formal

La **regresión lineal** es la aplicación estadística del método de mínimos cuadrados para modelar la relación entre una variable dependiente $y$ y una o más variables independientes $x$.

### Modelo de Regresión Lineal Simple

$$y_i = \beta_0 + \beta_1 x_i + \varepsilon_i$$

Donde:
- $\beta_0$ = intercepto poblacional (estimado por $b$)
- $\beta_1$ = pendiente poblacional (estimada por $m$)
- $\varepsilon_i$ = error aleatorio o residual

### Estimadores de Mínimos Cuadrados Ordinarios (MCO)

Los estimadores $\hat{\beta}_0$ y $\hat{\beta}_1$ que minimizan $\sum \varepsilon_i^2$:

$$\hat{\beta}_1 = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^2}$$

$$\hat{\beta}_0 = \bar{y} - \hat{\beta}_1\bar{x}$$

### Propiedades de los Estimadores MCO

1. **Linealidad**: Los estimadores son combinaciones lineales de las $y_i$
2. **Insesgadez**: $E[\hat{\beta}] = \beta$ (el valor esperado es el verdadero parámetro)
3. **Eficiencia**: Varianza mínima entre todos los estimadores lineales insesgados (Teorema de Gauss-Markov)

### Coeficiente de Determinación $R^2$

Mide la proporción de variabilidad explicada por el modelo:

$$R^2 = 1 - \frac{\sum(y_i - \hat{y}_i)^2}{\sum(y_i - \bar{y})^2} = 1 - \frac{SS_{res}}{SS_{tot}}$$

donde $0 \leq R^2 \leq 1$. Valores cercanos a 1 indican mejor ajuste.

## (c) Regresión Polinómica

### Extensión del Modelo Lineal

Cuando la relación entre variables no es lineal, extendemos el modelo a **polinomios de grado $p$**:

$$y = a_0 + a_1 x + a_2 x^2 + \cdots + a_p x^p = \sum_{j=0}^{p} a_j x^j$$

### Función de Error

$$S(a_0, a_1, \ldots, a_p) = \sum_{i=1}^{n} \left(y_i - \sum_{j=0}^{p} a_j x_i^j\right)^2$$

### Sistema de Ecuaciones Normales (Forma Matricial)

Minimizando mediante derivadas parciales $\frac{\partial S}{\partial a_k} = 0$ para $k = 0, 1, \ldots, p$:

$$\sum_{j=0}^{p} a_j \sum_{i=1}^{n} x_i^{j+k} = \sum_{i=1}^{n} y_i x_i^k \quad \text{para } k = 0, 1, \ldots, p$$

### Formulación Matricial Elegante

Definiendo la **matriz de Vandermonde** $\mathbf{X}$:

$$\mathbf{X} = \begin{bmatrix} 
1 & x_1 & x_1^2 & \cdots & x_1^p \\
1 & x_2 & x_2^2 & \cdots & x_2^p \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_n & x_n^2 & \cdots & x_n^p
\end{bmatrix}, \quad 
\mathbf{a} = \begin{bmatrix} a_0 \\ a_1 \\ \vdots \\ a_p \end{bmatrix}, \quad
\mathbf{y} = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{bmatrix}$$

El sistema se reduce a:

$$\mathbf{X}^T\mathbf{X}\mathbf{a} = \mathbf{X}^T\mathbf{y}$$

### Solución Matricial

$$\mathbf{a} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}$$

&gt; **Nota importante**: Aunque el modelo es **no lineal en $x$**, sigue siendo **lineal en los parámetros** $a_j$, por lo que seguimos en el dominio de mínimos cuadrados lineales.

### Consideraciones Prácticas

- **Grado del polinomio**: Mayor $p$ implica mejor ajuste pero riesgo de **sobreajuste** (overfitting)
- **Condicionamiento**: Matrices de Vandermonde mal condicionadas para $p$ grande
- **Alternativas**: Usar polinomios ortogonales o splines cúbicos para casos complejos

## (d) Ecuaciones Paramétricas y su Relación con Mínimos Cuadrados

### ¿Qué son las Ecuaciones Paramétricas?

Las **ecuaciones paramétricas** describen una curva donde las coordenadas $x$ e $y$ se expresan como funciones de un **parámetro independiente** $t$:

$$\begin{cases}
x = x(t) \\
y = y(t)
\end{cases} \quad \text{con } t \in [t_{min}, t_{max}]$$

### Ajuste de Curvas Paramétricas por Mínimos Cuadrados

Cuando los datos $(x_i, y_i)$ provienen de una trayectoria curva (círculos, elipses, trayectorias físicas), el ajuste paramétrico es más natural que $y = f(x)$.

#### Caso 1: Ajuste de Círculo

Ecuación paramétrica de un círculo de radio $R$ centrado en $(a,b)$:

$$\begin{cases}
x(t) = a + R\cos(t) \\
y(t) = b + R\sin(t)
\end{cases}$$

**Problema de mínimos cuadrados**: Dados puntos $(x_i, y_i)$, encontrar $a, b, R$ que minimicen:

$$S = \sum_{i=1}^{n} \left[(x_i - a)^2 + (y_i - b)^2 - R^2\right]^2$$

#### Caso 2: Ajuste de Elipse

$$\begin{cases}
x(t) = a + A\cos(t)\cos(\phi) - B\sin(t)\sin(\phi) \\
y(t) = b + A\cos(t)\sin(\phi) + B\sin(t)\cos(\phi)
\end{cases}$$

donde $(a,b)$ es el centro, $A, B$ los semiejes, y $\phi$ el ángulo de rotación.

### Formulación General del Ajuste Paramétrico

Dado un modelo vectorial $\mathbf{r}(t; \mathbf{p})$ con parámetros $\mathbf{p} = (p_1, \ldots, p_m)$:

$$\min_{\mathbf{p}} \sum_{i=1}^{n} \|\mathbf{r}_i - \mathbf{r}(t_i; \mathbf{p})\|^2$$

donde $\|\cdot\|$ denota la **distancia euclidiana**.

### Ventajas del Enfoque Paramétrico

| Aspecto | Función $y=f(x)$ | Ecuaciones Paramétricas |
|---------|------------------|---------------------------|
| **Curvas cerradas** | No puede representar (ej: círculo completo) | ✓ Representa naturalmente |
| **Multivaluadas** | Problemas con $x$ repetidos | ✓ Sin restricciones |
| **Movimiento físico** | No incluye tiempo explícitamente | ✓ $t$ representa tiempo |
| **Optimización** | Lineal en parámetros (simplificada) | Generalmente no lineal |

### Aplicación: Suavizado de Trayectorias

Para datos ruidosos de una trayectoria $(x_i, y_i)$ con tiempos $t_i$:

1. **Ajuste independiente**:
   - Ajustar $x(t) \approx \sum a_j \phi_j(t)$ por mínimos cuadrados
   - Ajustar $y(t) \approx \sum b_j \phi_j(t)$ por mínimos cuadrados
   
   donde $\phi_j(t)$ pueden ser polinomios, splines o funciones de base.

2. **Solución matricial separada**:

$$\mathbf{a} = (\mathbf{\Phi}^T\mathbf{\Phi})^{-1}\mathbf{\Phi}^T\mathbf{x}, \quad \mathbf{b} = (\mathbf{\Phi}^T\mathbf{\Phi})^{-1}\mathbf{\Phi}^T\mathbf{y}$$

con $\mathbf{\Phi}_{ij} = \phi_j(t_i)$.

### Conclusión Conceptual

&gt; Las ecuaciones paramétricas extienden el poder de los mínimos cuadrados más allá de funciones $y = f(x)$, permitiendo modelar **curvas complejas, trayectorias temporales y formas geométricas** mediante la descomposición en componentes independientes, cada una ajustable por el mismo principio de minimización del error cuadrático.


### Ecuaciones Fundamentales Recurrentes

| Contexto | Ecuación Normal |
|------------|-----------------|
| Lineal simple | $n$ ecuaciones, 2 incógnitas $(m,b)$ |
| Polinómica | $(p+1)$ ecuaciones, $(p+1)$ incógnitas |
| Paramétrica | 2 sistemas independientes (para $x(t)$ e $y(t)$) |

### Principio Unificador

> Todos los casos derivan del mismo principio: **la derivada del error cuadrático respecto a cada parámetro debe anularse**, generando un sistema de ecuaciones lineales en los parámetros del modelo (aunque el modelo sea no lineal en las variables independientes).
