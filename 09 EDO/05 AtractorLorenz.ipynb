# Función matemática que parece una arroba que se junta con otra arroba invertida
La función matemática que describes, que al graficarse parece una arroba (@) junta con otra invertida, es la **gráfica de una función y su función inversa** ($f(x)$ y $f^{-1}(x)$) reflejadas respecto a la línea diagonal $y = x$.

## Descripción de la Gráfica

- **La "Arroba" ($f(x)$)**: Generalmente, funciones exponenciales ($e^x$) o cúbicas ($x^3$) curvadas que se aproximan a un eje forman una curva parecida a la espiral interna de una arroba.

- **La "Arroba Invertida" ($f^{-1}(x)$)**: Es la reflexión espejo de la primera función. Cuando graficas una función y su inversa, la simetría respecto a la recta $y = x$ crea este efecto, donde una curva es el "espejo" de la otra, pareciendo un símbolo @ y un símbolo @ al revés unidos.

## Ejemplos Comunes

1. **Función exponencial y logarítmica**: $f(x) = e^x$ y $g(x) = \ln(x)$.

2. **Función cúbica y raíz cúbica**: $f(x) = x^3$ y $g(x) = \sqrt[3]{x}$.

Si graficas $y = x^3$ y $y = \sqrt[3]{x}$ juntas en una calculadora gráfica como Desmos, obtendrás esa forma de "doble arroba" o "arroba invertida" simétrica.



# Parece una espiral
Si lo que buscas es una sola curva continua que se enrosca sobre sí misma formando dos espirales (una "derecha" y otra "invertida"), lo más probable es que te refieras a la **Clotoide** (también conocida como la **Espiral de Cornu** o de Euler).

## La Clotoide (Espiral de Cornu)

Esta función es fascinante porque empieza en el centro, se curva hacia un lado formando una "arroba" y, hacia el otro lado, se curva exactamente igual pero de forma invertida.

- **Forma visual**: Parece una letra "S" infinita donde los extremos se enrollan sobre sí mismos infinitamente hacia dos puntos focales.

- **Uso en la vida real**: Se usa en el **diseño de carreteras y vías de tren**. Es la curva que permite que un conductor gire el volante de forma constante y suave al entrar en una curva, evitando cambios bruscos de fuerza centrífuga.

## Definición Matemática

Se define mediante las **Integrales de Fresnel**, donde la curvatura aumenta linealmente con la distancia recorrida:

$$S(x) = \int_0^x \sin(t^2) \, dt$$

$$C(x) = \int_0^x \cos(t^2) \, dt$$

Al graficar estas dos funciones como coordenadas $(x, y)$, obtienes esa doble espiral perfecta.

## Otras opciones similares:

- **Lemniscata de Bernoulli**: Se parece más a un símbolo de infinito ($\infty$), pero tiene esa simetría de dos lazos encontrados.

- **Espiral de Fermat**: Es una espiral doble que nace del origen; una rama gira hacia afuera y la otra hace lo mismo en sentido opuesto ($r^2 = a^2\theta$).

> **Nota curiosa**: Si ves la Clotoide, notarás que es la forma más elegante de unir una línea recta con un círculo perfecto.

# O un planeta siendo tragado por un agujero negro

## La Doble Espiral (Planeta siendo "devorado" por ambos lados)

Si ves dos "arrobas" o espirales encontradas, como si el planeta se estuviera dividiendo en dos chorros de materia, matemáticamente podrías estar viendo una **Espiral de Fermat** ($r^2 = a^2\theta$).

Esta función crea dos brazos simétricos que nacen del mismo centro:

1. Un brazo gira en un sentido (la "arroba" normal).

2. El otro brazo gira en el sentido opuesto (la "arroba" invertida).

Visualmente, parece una galaxia o un sistema donde la materia está siendo succionada desde dos puntos opuestos hacia un centro común.

Esta descripción es muy evocadora y apunta directamente a un fenómeno de la astrofísica llamado **Espaguetificación** (o efecto de marea), pero si buscamos la función matemática que dibuja ese "chorro" de materia siendo succionado y enrollado, estamos hablando de una **Espiral de Lituus**.

## La función del "Accretimiento" (Espiral de Lituus)

Esta función representa perfectamente a un objeto (el planeta) siendo estirado y curvado hacia un centro denso (el agujero negro). A medida que se acerca al centro, la curva se vuelve más cerrada y rápida, pareciendo una "arroba" extremadamente estirada.<br>

La ecuación en coordenadas polares es:

$$r^2 = \frac{a^2}{\theta}$$

- **Por qué se parece**: A diferencia de otras espirales que crecen hacia afuera, el **Lituus** tiene una asíntota en el eje $x$. Es decir, viene desde muy lejos como una línea casi recta (el planeta acercándose) y de repente comienza a enroscarse violentamente hacia el centro.

# Mencionaste un atractor
¡Exacto! Esa es la pieza del rompecabezas que faltaba. Lo que tienes en mente es probablemente el **Atractor de Lorenz**.
Cuando mencionas el "planeta siendo tragado por un agujero negro" y las "dos arrobas", el Atractor de Lorenz encaja perfectamente porque visualmente parece una **mariposa** o **dos espirales infinitas** que se envuelven una a la otra sin tocarse nunca.

## ¿Qué es el Atractor de Lorenz?

Es el ejemplo más famoso de la **Teoría del Caos**. Fue descubierto por Edward Lorenz mientras intentaba modelar el clima. Representa un sistema donde pequeñas variaciones iniciales producen resultados totalmente distintos (el famoso **"Efecto Mariposa"**).

## ¿Por qué encaja con tu descripción?

1. **Las dos arrobas**: La gráfica consiste en una trayectoria que da vueltas en una espiral (una "arroba"), y de repente "salta" hacia la otra espiral (la arroba invertida).

2. **El agujero negro**: Cada espiral gira alrededor de un punto central llamado **"punto de equilibrio"**. El trazo parece ser succionado perpetuamente hacia esos centros, pero nunca llega a tocarlos.

3. **El planeta siendo estirado**: La forma en que las líneas se separan y se enrollan recuerda mucho a la materia siendo estirada por fuerzas gravitatorias extremas.

## La Matemática detrás (Ecuaciones de Lorenz)

No es una sola función simple, sino un **sistema de tres ecuaciones diferenciales**:

$$\frac{dx}{dt} = \sigma(y - x)$$

$$\frac{dy}{dt} = x(\rho - z) - y$$

$$\frac{dz}{dt} = xy - \beta z$$

> **Curiosidad**: Aunque parece un dibujo en 2D, el atractor es en realidad un objeto **tridimensional**. Lo que vemos como "dos arrobas" es la sombra o proyección de una trayectoria que nunca se cruza consigo misma.

### Los Tres Parámetros Adimensionales

| Parámetro | Nombre | Definición Física | Valor Clásico |
|-----------|--------|-------------------|---------------|
| $\sigma$ | **Número de Prandtl** | $\sigma = \frac{\nu}{\kappa}$ | 10 |
| $\rho$ | **Número de Rayleigh reducido** | $\rho = \frac{Ra}{Ra_c}$ | 28 |
| $\beta$ | **Parámetro geométrico** | $\beta = \frac{4}{1 + a^2}$ | 8/3 |

donde:
- $\nu$ = viscosidad cinemática del fluido
- $\kappa$ = difusividad térmica  
- $Ra$ = número de Rayleigh (fuerza de convección)
- $Ra_c$ = número de Rayleigh crítico (inicio de convección)
- $a$ = número de onda horizontal del patrón de convección

---

### $\sigma$ — Número de Prandtl

$$\sigma = \frac{\text{Difusión de momentum (viscosidad)}}{\text{Difusión de calor}} = \frac{\nu}{\kappa}$$

#### Interpretación Física

| Valor de $\sigma$ | Significado | Ejemplo |
|-------------------|-------------|---------|
| $\sigma \ll 1$ | El calor se difunde mucho más rápido que el momentum | Mercurio líquido ($\sigma \approx 0.025$) |
| $\sigma \approx 1$ | Procesos comparables | Gases ($\sigma \approx 0.7$) |
| $\sigma \gg 1$ | El momentum se difunde mucho más rápido | Aceite de motor ($\sigma \approx 1000$), manto terrestre |
| $\sigma = 10$ | Valor usado por Lorenz | Agua a temperatura ambiente |

#### Efecto en el Caos

- $\sigma$ controla la **escala de tiempo** de la dinámica
- Para $\sigma < 1$, el sistema puede exhibir comportamiento diferente
- El valor $\sigma = 10$ asegura que la inercia viscosa domina sobre la difusión térmica

---

### $\rho$ — Número de Rayleigh Reducido

$$\rho = \frac{Ra}{Ra_c} = \frac{\text{Fuerza de convección}}{\text{Fuerza de disipación crítica}}$$

#### Interpretación Física

| Rango de $\rho$ | Régimen Físico | Comportamiento |
|-----------------|----------------|----------------|
| $0 < \rho < 1$ | Conducción pura | Estado estacionario trivial (sin movimiento) |
| $\rho = 1$ | Umbral de convección | Inestabilidad, aparecen rollos de convección |
| $1 < \rho < 13.926$ | Convección estable | Dos estados estacionarios estables |
| $13.926 < \rho < 24.06$ | Biestabilidad | Coexisten estados estacionarios y caos |
| $24.06 < \rho < 24.74$ | Ventana de caos intermitente | Caos y periodicidad alternan |
| $\rho > 24.74$ | **Caos determinista** | Atractor extraño de Lorenz |

#### Valor Crítico: $\rho = 28$

Lorenz eligió $\rho = 28$ porque está **profundamente en el régimen caótico**, lejos de las transiciones complejas cerca de $\rho \approx 24$.

---

### $\beta$ — Parámetro Geométrico de Aspecto

$$\beta = \frac{4}{1 + a^2}$$

donde $a = \frac{\pi h}{L}$ es el número de onda adimensional.

#### Interpretación Física

- $\beta$ relaciona la **altura** $h$ con el **ancho** $L$ de la celda de convección
- Valor $\beta = 8/3 \approx 2.667$ corresponde a $a = 1/2$, es decir, celdas de convección con **relación de aspecto 2:1**

#### Efecto en la Geometría del Atractor

| Valor de $\beta$ | Característica del Atractor |
|------------------|----------------------------|
| $\beta \to 0$ ($a \to \infty$) | Celdas muy anchas, dinámica degenerada |
| $\beta = 8/3$ | **Atractor clásico de mariposa** |
| $\beta \to 4$ ($a \to 0$) | Celdas muy altas, comportamiento rígido |

---

### Valores Clásicos y sus Propiedades

$$\sigma = 10, \quad \rho = 28, \quad \beta = \frac{8}{3}$$



# El atractor de lorenz puede ser simulado en python
¡Totalmente! Es uno de los experimentos clásicos para visualizar la **Teoría del Caos**. Para simularlo, se suelen usar las librerías `numpy` (para los cálculos) y `matplotlib` (para los gráficos 3D).<br>

Aquí tienes un ejemplo rápido de cómo se hace usando el integrador `odeint` de la librería SciPy.org:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Ecuaciones diferenciales de Lorenz
def lorenz(state, t, sigma, rho, beta):
    x, y, z = state
    return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]

# Parámetros estándar para ver las "dos arrobas" (caos)
sigma, rho, beta = 10.0, 28.0, 8.0/3.0
state0 = [1.0, 1.0, 1.0]  # Punto inicial
t = np.arange(0.0, 40.0, 0.01)  # Tiempo de simulación

# Resolver el sistema
states = odeint(lorenz, state0, t, args=(sigma, rho, beta))

# Graficar en 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(states[:, 0], states[:, 1], states[:, 2], lw=0.5, color='purple')
plt.show()

## ¿Por qué es genial simularlo?

1. **Sensibilidad inicial**: Si cambias el `state0` de `[1.0, 1.0, 1.0]` a `[1.0, 1.0, 1.01]`, verás que al principio las líneas van juntas, pero pronto se separan y terminan en "arrobas" distintas.

2. **Interactividad**: Si usas herramientas como Plotly Python, puedes rotar el gráfico para ver cómo el "planeta" (la trayectoria) es succionado de un ala a la otra.

3. **Animación**: Puedes usar el módulo `animation` de Matplotlib para ver el trazo dibujándose en tiempo real, lo que da esa sensación de materia fluyendo hacia un centro gravitatorio.

# Estaba pensando escribir una animación en raylib con una esfera pequeña en el centro de la espiral pequeña y pasando pequeñas partículas a una esfera negra más grande coronando el agujero negro
Estaba pensando escribir una animación en raylib con una esfera pequeña en el centro de la espiral pequeña y pasando pequeñas partículas a una esfera negra más grande coronando el agujero negro

## 1. El Motor de Movimiento (Lorenz)

En lugar de pre-calcular todo, actualiza la posición de las partículas en el `update` usando el método de Euler (es más rápido para tiempo real):

```cpp
// Dentro del Loop:
float dx = (sigma * (py - px)) * dt;
float dy = (px * (rho - pz) - py) * dt;
float dz = (px * py - beta * pz) * dt;
px += dx; py += dy; pz += dz;
```

## 1. La Fuerza de Lorentz (No confundir con Lorenz)

Mientras que el Atractor de Lorenz dicta la "forma de arroba", la **Fuerza de Lorentz** dicta cómo se mueve una partícula cargada en un campo electromagnético. La fórmula es:

$$\mathbf{F} = q(\mathbf{E} + \mathbf{v} \times \mathbf{B})$$

- **$\mathbf{E}$ (Campo Eléctrico):** Podría ser una fuerza radial que atrae las partículas hacia el centro del agujero negro.

- **$\mathbf{B}$ (Campo Magnético):** Es lo que genera las espirales. En un agujero negro, las líneas de campo magnético salen de los polos, creando **jets relativistas** (chorros de partículas disparados hacia arriba y abajo).

## 2. Implementación en tu simulación

Para que las partículas en raylib se muevan de forma orgánica entre la esfera pequeña y el agujero negro, puedes usar un sistema de **"Fuerzas Combinadas"**:

- **El Atractor como "Guía":** Usa las ecuaciones de Lorenz para definir el "viento" o flujo base del espacio.

- **Magnetismo para el "Brillo":** Haz que las partículas más cercanas al agujero negro giren más rápido (fuerza centrípeta magnética). Esto creará el **disco de acreción** (ese anillo plano y brillante que se ve en las fotos reales de agujeros negros).

```cpp
// En el Update de cada partícula
Vector3 getLorenzVelocity(Vector3 p) {
    return (Vector3){
        sigma * (p.y - p.x),
        p.x * (rho - p.z) - p.y,
        p.x * p.y - beta * p.z
    };
}
// Luego sumas esa velocidad a la posición de la partícula
particle_pos = Vector3Add(particle_pos,
    Vector3Scale(getLorenzVelocity(particle_pos), dt));

# Y si en lugar de Euler RK4?

¡Excelente elección! Si vas a simular el **Atractor de Lorenz** o campos de fuerzas, **Runge-Kutta de 4º Orden (RK4)** es el estándar de oro.

El método de Euler (que usamos antes) asume que la velocidad es constante durante todo el paso de tiempo ($dt$). En curvas cerradas como las de una "arroba" o cerca de un agujero negro, Euler comete un error acumulativo enorme: las partículas saldrían disparadas hacia el infinito (divergencia) o "se saldrían de la pista".

RK4 analiza la pendiente en cuatro puntos diferentes del intervalo para predecir la trayectoria con muchísima más precisión.

## Implementación de RK4 para el Atractor de Lorenz en C++

Aquí tienes cómo estructurar la función de actualización para tus partículas en **raylib**:

```cpp
#include "raylib.h"
#include "raymath.h"

// Parámetros de Lorenz
const float sigma = 10.0f;
const float rho = 28.0f;
const float beta = 8.0f / 3.0f;

// La función que define las derivadas (el "campo de velocidades")
Vector3 LorenzDeriv(Vector3 p) {
    return {
        sigma * (p.y - p.x),
        p.x * (rho - p.z) - p.y,
        p.x * p.y - beta * p.z
    };
}

// Implementación de RK4
Vector3 UpdateRK4(Vector3 p, float dt) {
    Vector3 k1 = LorenzDeriv(p);
    
    Vector3 p2 = Vector3Add(p, Vector3Scale(k1, dt * 0.5f));
    Vector3 k2 = LorenzDeriv(p2);
    
    Vector3 p3 = Vector3Add(p, Vector3Scale(k2, dt * 0.5f));
    Vector3 k3 = LorenzDeriv(p3);
    
    Vector3 p4 = Vector3Add(p, Vector3Scale(k3, dt));
    Vector3 k4 = LorenzDeriv(p4);
    
    // Combinación ponderada: p + (dt/6) * (k1 + 2*k2 + 2*k3 + k4)
    Vector3 final_k = Vector3Scale(
        Vector3Add(k1, Vector3Add(Vector3Scale(k2, 2.0f),
        Vector3Add(Vector3Scale(k3, 2.0f), k4)),
        dt / 6.0f
    );
    
    return Vector3Add(p, final_k);
}
```

# Código integrado

```cpp
#include "raylib.h"
#include "raymath.h"
#include <vector>
#include <deque>

// --- Configuración del Atractor ---
const float sigma = 10.0f;
const float rho = 28.0f;
const float beta = 8.0f / 3.0f;
const float dt = 0.01f;              // Paso de tiempo para RK4
const int MAX_TRAIL = 20;            // Longitud del rastro de cada partícula
const int NUM_PARTICLES = 100;       // Cantidad de "materia" fluyendo

struct Particle {
    Vector3 position;
    std::deque<Vector3> trail;
    Color color;
    
    void Update() {
        // --- Implementación de RK4 ---
        auto Deriv = [](Vector3 p) -> Vector3 {
            return {
                sigma * (p.y - p.x),
                p.x * (rho - p.z) - p.y,
                p.x * p.y - beta * p.z
            };
        };
        
        Vector3 k1 = Deriv(position);
        Vector3 k2 = Deriv(Vector3Add(position, Vector3Scale(k1, dt * 0.5f)));
        Vector3 k3 = Deriv(Vector3Add(position, Vector3Scale(k2, dt * 0.5f)));
        Vector3 k4 = Deriv(Vector3Add(position, Vector3Scale(k3, dt)));
        
        Vector3 velocity = Vector3Scale(
            Vector3Add(Vector3Add(k1, Vector3Scale(k2, 2.0f)),
            Vector3Add(Vector3Scale(k3, 2.0f), k4)),
            dt / 6.0f
        );
        
        position = Vector3Add(position, velocity);
        
        // Actualizar rastro
        trail.push_front(position);
        if (trail.size() > MAX_TRAIL) trail.pop_back();
    }
};

while (!WindowShouldClose()) {
    // Actualización física
    for (auto &p : particles) p.Update();
    UpdateCamera(&camera, CAMERA_ORBITAL);
    
    BeginDrawing();
        ClearBackground(BLACK);
        BeginMode3D(camera);
        
            // 1. Dibujar el "Agujero Negro" y la "Estrella" en los centros
            //    de los atractores
            DrawSphere(Vector3){ 8.48f, 8.48f, 27.0f }, 2.0f, Fade(PURPLE, 0.3f));
            DrawSphere(Vector3){-8.48f, -8.48f, 27.0f }, 1.5f, YELLOW);
            
            // 2. Dibujar trayectorias de partículas
            for (const auto &p : particles) {
                for (size_t i = 0; i < p.trail.size() - 1; i++) {
                    float alpha = 1.0f - ((float)i / MAX_TRAIL);
                    DrawLine3D(p.trail[i], p.trail[i+1], Fade(p.color, alpha));
                }
            }
            
            DrawGrid(10, 10.0f);
        EndMode3D();
        
        DrawFPS(10, 10);
        DrawText("RK4: Estabilidad en transferencia de masa", 10, 30, 20, RAYWHITE);
    EndDrawing();
}

CloseWindow();
return 0;

```

```bash
g++ main.cpp -lraylib -lGL -lm -lpthread -ldl -lrt -lX11  # (en Linux)

# Había considerado rotación en sentido inverso de las manecilla de reloj del planeta y el agujero en sentido de las manecillas de reloj

Esta es una idea brillante desde el punto de vista de la dinámica de sistemas. En física, esto se asemeja a un **sistema binario con espines opuestos**, lo que genera una zona de cizallamiento (torsión) en el puente de materia entre ambos.

Para implementar esto en tu simulación de Raylib, no necesitamos cambiar las ecuaciones de Lorenz (que ya son intrínsecamente rotacionales), sino **ajustar el campo de velocidades** o aplicar una **transformación de coordenadas local** cerca de cada cuerpo.

Aquí tienes cómo integrar esas rotaciones opuestas en el motor de partículas:

## 1. Definir los centros de rotación

Primero, identificamos los focos del atractor (donde están tus esferas):

- **Foco A (Planeta):** $C_1 = (-\sqrt{\beta(\rho-1)}, -\sqrt{\beta(\rho-1)}, \rho-1)$

- **Foco B (Agujero Negro):** $C_2 = (\sqrt{\beta(\rho-1)}, \sqrt{\beta(\rho-1)}, \rho-1)$

## 2. Aplicar "Torque" diferencial en el RK4

Dentro de tu función `LorenzDeriv`, podemos añadir un término de rotación extra que dependa de la distancia a cada centro.

```cpp
Vector3 LorenzDerivConSpin(Vector3 p) {
    // Velocidad base de Lorenz
    Vector3 v = {
        sigma * (p.y - p.x),
        p.x * (rho - p.z) - p.y,
        p.x * p.y - beta * p.z
    };
    
    // Vector desde los centros a la partícula
    Vector3 distA = Vector3Subtract(p, centroPlaneta);
    Vector3 distB = Vector3Subtract(p, centroAgujero);
    
    float fuerzaSpin = 5.0f;  // Intensidad de la rotación extra
    
    if (Vector3Length(distA) < 15.0f) {
        // Rotación ANTI-HORARIA (Planeta)
        // Producto cruz con el eje Z (0,0,1) genera giro en el plano XY
        Vector3 spinA = { -distA.y, distA.x, 0 };
        v = Vector3Add(v, Vector3Scale(Vector3Normalize(spinA), fuerzaSpin));
    }
    
    if (Vector3Length(distB) < 15.0f) {
        // Rotación HORARIA (Agujero Negro)
        Vector3 spinB = { distB.y, -distB.x, 0 };
        v = Vector3Add(v, Vector3Scale(Vector3Normalize(spinB), fuerzaSpin));
    }
    
    return v;
}

```cpp
Vector3 LorenzDerivConSpin(Vector3 p) {
    // 1. Velocidad base del Atractor de Lorenz (La estructura macro)
    Vector3 v = {
        sigma * (p.y - p.x),
        p.x * (rho - p.z) - p.y,
        p.x * p.y - beta * p.z
    };
    
    // Focos teóricos (donde ocurren las "arrobas")
    float b_rho = sqrtf(beta * (rho - 1.0f));
    Vector3 focoPlaneta = { -b_rho, -b_rho, rho - 1.0f };
    Vector3 focoAgujero = { b_rho, b_rho, rho - 1.0f };
    
    // 2. Parámetros de los espines
    float radioInfluencia = 12.0f;  // Qué tanto se extiende el giro
    float fuerzaSpin = 15.0f;       // Intensidad de la rotación
    
    // --- Rotación ANTI-HORARIA (Planeta) ---
    Vector3 distA = Vector3Subtract(p, focoPlaneta);
    float distALen = Vector3Length(distA);
    // Factor Gaussiano: 1.0 en el centro, cerca de 0 lejos
    float factorA = expf(-(distALen * distALen) / (2.0f * radioInfluencia * radioInfluencia));
    Vector3 torqueA = { -distA.y, distA.x, 0 };  // Giro Anti-horario en XY
    v = Vector3Add(v, Vector3Scale(Vector3Normalize(torqueA), fuerzaSpin * factorA));
    
    // --- Rotación HORARIA (Agujero Negro) ---
    Vector3 distB = Vector3Subtract(p, focoAgujero);
    float distBLen = Vector3Length(distB);
    float factorB = expf(-(distBLen * distBLen) / (2.0f * radioInfluencia * radioInfluencia));
    Vector3 torqueB = { distB.y, -distB.x, 0 };  // Giro Horario en XY
    v = Vector3Add(v, Vector3Scale(Vector3Normalize(torqueB), fuerzaSpin * factorB));
    
    return v;
}

## 1. El Campo de Velocidades Refinado

Implementaremos un **Gradiente de Influencia**. Definiremos una zona muerta en el centro del sistema para que las partículas recuperen la trayectoria natural de Lorenz antes de ser capturadas por el siguiente giro.

```cpp
Vector3 CalcularVelocidadRefinada(Vector3 p, float dt) {
    // 1. Vector base de Lorenz (La estructura macroscópica)
    Vector3 v_lorenz = {
        sigma * (p.y - p.x),
        p.x * (rho - p.z) - p.y,
        p.x * p.y - beta * p.z
    };
    
    // Focos del Atractor (Puntos L1 y L2 del sistema)
    float b_rho = sqrtf(beta * (rho - 1.0f));
    Vector3 centroPlaneta = { -b_rho, -b_rho, rho - 1.0f };
    Vector3 centroAgujero = { b_rho, b_rho, rho - 1.0f };
    
    // 2. Cálculo de Influencia Relativa
    Vector3 r_planeta = Vector3Subtract(p, centroPlaneta);
    Vector3 r_agujero = Vector3Subtract(p, centroAgujero);
    
    float d_p = Vector3Length(r_planeta);
    float d_a = Vector3Length(r_agujero);
    
    // Parámetros de diseño físico
    float R_influencia = 15.0f;   // Radio de captura
    float intensidad_giro = 25.0f;
    
    // 3. Suavizado de transición (Smoothstep invertido)
    // Esto asegura que la fuerza de rotación sea 0 fuera del radio R
    auto SmoothWeight = [](float d) -> float {
        float x = fmaxf(0.0f, 1.0f - (d / R_influencia));
        return x * x * (3.0f - 2.0f * x);  // Función Smoothstep
    };
    
    float w_p = SmoothWeight(d_p);
    float w_a = SmoothWeight(d_a);
    
    // 4. Vectores de Rotación Pura (Perpendiculares al radio y al eje Z)
    // Giro Anti-horario (Planeta): (0,0,1) x r_p -> {-y, x, 0}
    Vector3 v_spin_p = { -r_planeta.y, r_planeta.x, 0 };
    // Giro Horario (Agujero): (0,0,1) x r_a -> {y, -x, 0}
    Vector3 v_spin_a = { r_agujero.y, -r_agujero.x, 0 };
    
    // 5. Integración Final (Mezcla ponderada)
    Vector3 v_final = v_lorenz;
    
    if (w_p > 0) v_final = Vector3Add(v_final,
        Vector3Scale(Vector3Normalize(v_spin_p), intensidad_giro * w_p));
    if (w_a > 0) v_final = Vector3Add(v_final,
        Vector3Scale(Vector3Normalize(v_spin_a), intensidad_giro * w_a));
    
    return v_final;
}

# Código completo
```cpp
#include "raylib.h"
#include "raymath.h"
#include <cmath>
#include <vector>
#include <deque>

// ============================================
// CONFIGURACION
// ============================================
const int MAX_PARTICULAS = 500;
const int MAX_TRAIL = 50;

float sigma = 10.0f;
float rho = 28.0f;
float beta = 8.0f / 3.0f;
float dt = 0.01f;
float factorLorenz = 0.04f;
float fuerzaSpin = 8.0f;

// ============================================
// CÁMARA MANUAL
// ============================================
struct CamaraControl {
    float rotX = 30.0f;
    float rotY = 45.0f;
    float distancia = 70.0f;
    Vector3 objetivo = { 0, 0, 25 };
    bool arrastrando = false;
    Vector2 posAnterior;

    void actualizar() {
        if (IsMouseButtonPressed(MOUSE_LEFT_BUTTON)) {
            arrastrando = true;
            posAnterior = GetMousePosition();
        }
        if (IsMouseButtonReleased(MOUSE_LEFT_BUTTON)) {
            arrastrando = false;
        }

        if (arrastrando) {
            Vector2 mouse = GetMousePosition();
            rotY += (mouse.x - posAnterior.x) * 0.5f;
            rotX -= (mouse.y - posAnterior.y) * 0.5f;

            if (rotX > 85.0f) rotX = 85.0f;
            if (rotX < -85.0f) rotX = -85.0f;
            posAnterior = mouse;
        }

        float rueda = GetMouseWheelMove();
        distancia -= rueda * 3.0f;
        distancia = fmaxf(20.0f, fminf(150.0f, distancia));
    }

    Camera3D obtenerCamara() {
        float rx = rotX * DEG2RAD;
        float ry = rotY * DEG2RAD;

        Camera3D cam = { 0 };
        cam.position = {
            objetivo.x + distancia * cosf(rx) * sinf(ry),
            objetivo.y + distancia * sinf(rx),
            objetivo.z + distancia * cosf(rx) * cosf(ry)
        };
        cam.target = objetivo;
        cam.up = { 0, 1, 0 };
        cam.fovy = 60.0f;
        cam.projection = CAMERA_PERSPECTIVE;
        return cam;
    }
};

// ============================================
// PARTÍCULA CON RK4 + TORQUE DIFERENCIAL
// ============================================
struct Particula {
    Vector3 pos;
    Vector3 vel;
    std::deque<Vector3> rastro;
    Color color;
    float vida;
    bool activa;
    int tipo; // 0 = flujo a agujero, 1 = órbita planeta

    Particula() : vida(0), activa(false), tipo(0) {}

    void reset(Vector3 p, Vector3 v, Color c, int t) {
        pos = p;
        vel = v;
        color = c;
        vida = 0;
        activa = true;
        tipo = t;
        rastro.clear();
        rastro.push_back(p);
    }

    Vector3 lorenz(Vector3 p) {
        return {
            sigma * (p.y - p.x),
            p.x * (rho - p.z) - p.y,
            p.x * p.y - beta * p.z
        };
    }

    void calcularFuerzas(Vector3 p, Vector3& fLorenz, Vector3& fGrav,
        Vector3& fSpin, Vector3 planeta, Vector3 agujero) {

        fLorenz = lorenz(p);

        Vector3 haciaAgujero = Vector3Subtract(agujero, p);
        float distAgujero = Vector3Length(haciaAgujero);
        Vector3 dirAgujero = Vector3Normalize(haciaAgujero);

        Vector3 haciaPlaneta = Vector3Subtract(planeta, p);
        float distPlaneta = Vector3Length(haciaPlaneta);

        fGrav = { 0, 0, 0 };

        // Atracción al agujero (más débil para tipo 1)
        float masaAgujero = (tipo == 0) ? 500.0f : 150.0f;
        if (distAgujero > 1.0f) {
            fGrav = Vector3Add(fGrav,
                Vector3Scale(dirAgujero, masaAgujero / (distAgujero * distAgujero + 10.0f)));
        }

        // Atracción al planeta (más fuerte para tipo 1)
        float masaPlaneta = (tipo == 0) ? 50.0f : 200.0f;
        if (distPlaneta > 0.5f) {
            Vector3 dirPlaneta = Vector3Normalize(haciaPlaneta);
            fGrav = Vector3Add(fGrav,
                Vector3Scale(dirPlaneta, masaPlaneta / (distPlaneta * distPlaneta)));
        }

        // TORQUE DIFERENCIAL
        fSpin = { 0, 0, 0 };

        // Rotación ANTI-HORARIA cerca del planeta (más fuerte para tipo 1)
        if (distPlaneta < 25.0f) {
            Vector3 spinA = { -haciaPlaneta.y, haciaPlaneta.x, 0 };
            float factorA = 1.0f - (distPlaneta / 25.0f);
            float intensidad = (tipo == 0) ? fuerzaSpin : fuerzaSpin * 2.5f;
            fSpin = Vector3Add(fSpin,
                Vector3Scale(Vector3Normalize(spinA), intensidad * factorA));
        }

        // Rotación HORARIA cerca del agujero
        if (distAgujero < 20.0f) {
            Vector3 spinB = { haciaAgujero.y, -haciaAgujero.x, 0 };
            float factorB = 1.0f - (distAgujero / 20.0f);
            fSpin = Vector3Add(fSpin,
                Vector3Scale(Vector3Normalize(spinB), fuerzaSpin * factorB));
        }
    }

    void actualizar(Vector3 planeta, Vector3 agujero) {
        if (!activa) return;

        auto evaluar = [&](Vector3 p, Vector3 v, Vector3& dvdt) {
            Vector3 fLorenz, fGrav, fSpin;
            calcularFuerzas(p, fLorenz, fGrav, fSpin, planeta, agujero);

            Vector3 acc = Vector3Add(
                Vector3Scale(fLorenz, factorLorenz),
                Vector3Add(fGrav, fSpin)
            );
            dvdt = acc;
            return v;
            };

        Vector3 k1_p, k2_p, k3_p, k4_p;
        Vector3 k1_v, k2_v, k3_v, k4_v;
        Vector3 p_temp, v_temp;

        k1_p = evaluar(pos, vel, k1_v);

        p_temp = Vector3Add(pos, Vector3Scale(k1_p, dt * 0.5f));
        v_temp = Vector3Add(vel, Vector3Scale(k1_v, dt * 0.5f));
        k2_p = evaluar(p_temp, v_temp, k2_v);

        p_temp = Vector3Add(pos, Vector3Scale(k2_p, dt * 0.5f));
        v_temp = Vector3Add(vel, Vector3Scale(k2_v, dt * 0.5f));
        k3_p = evaluar(p_temp, v_temp, k3_v);

        p_temp = Vector3Add(pos, Vector3Scale(k3_p, dt));
        v_temp = Vector3Add(vel, Vector3Scale(k3_v, dt));
        k4_p = evaluar(p_temp, v_temp, k4_v);

        Vector3 suma_p = Vector3Add(
            Vector3Add(k1_p, Vector3Scale(k2_p, 2.0f)),
            Vector3Add(Vector3Scale(k3_p, 2.0f), k4_p)
        );
        Vector3 suma_v = Vector3Add(
            Vector3Add(k1_v, Vector3Scale(k2_v, 2.0f)),
            Vector3Add(Vector3Scale(k3_v, 2.0f), k4_v)
        );

        pos = Vector3Add(pos, Vector3Scale(suma_p, dt / 6.0f));
        vel = Vector3Add(vel, Vector3Scale(suma_v, dt / 6.0f));

        vel = Vector3Scale(vel, 0.99f);
        vida += dt;

        rastro.push_front(pos);
        if (rastro.size() > MAX_TRAIL) rastro.pop_back();
    }

    bool debeReset(Vector3 planeta, Vector3 agujero) {
        if (!activa) return true;
        float da = Vector3Distance(pos, agujero);
        float dp = Vector3Distance(pos, planeta);
        float tiempoMax = (tipo == 0) ? 15.0f : 25.0f; // Órbitas duran más
        return (da < 0.8f) || (dp > 100.0f) || (vida > tiempoMax) || (da > 150.0f);
    }

    void dibujar(Vector3 agujero) {
        if (!activa || rastro.size() < 2) return;

        float d = Vector3Distance(pos, agujero);
        float t = fminf(d / 30.0f, 1.0f);

        // Color diferente según tipo
        Color c;
        if (tipo == 0) {
            // Naranja a azul (flujo normal)
            c = Color{
                (unsigned char)(255 * t + 255 * (1.0f - t)),
                (unsigned char)(150 * t + 100 * (1.0f - t)),
                (unsigned char)(255 * (1.0f - t)),
                255
            };
        }
        else {
            // Verde (órbita planeta)
            c = Color{
                (unsigned char)(100 * t),
                255,
                (unsigned char)(100 * (1.0f - t)),
                255
            };
        }

        for (size_t i = 0; i + 1 < rastro.size(); i++) {
            float alpha = 1.0f - ((float)i / rastro.size());
            Color cc = c;
            cc.a = (unsigned char)(255 * alpha);
            DrawLine3D(rastro[i], rastro[i + 1], cc);
        }

        DrawSphere(pos, (tipo == 0) ? 0.15f : 0.2f, WHITE);
    }
};

// ============================================
// SISTEMA CON DOS TIPOS DE PARTÍCULAS
// ============================================
struct Sistema {
    std::vector<Particula> particulas;
    Vector3 planeta;
    Vector3 agujero;

    void init() {
        particulas.resize(MAX_PARTICULAS);
        calcularFocos();
        for (int i = 0; i < MAX_PARTICULAS; i++) spawn(i);
    }

    void calcularFocos() {
        float b = sqrtf(fmaxf(0.0f, beta * (rho - 1.0f)));
        planeta = { -b, -b, rho - 1.0f };
        agujero = { b, b, rho - 1.0f };
    }

    void spawn(int idx) {
        // 20% de partículas orbitan el planeta, 80% fluyen al agujero
        int tipo = (GetRandomValue(0, 100) < 20) ? 1 : 0;

        float ang = GetRandomValue(0, 360) * DEG2RAD;
        Vector3 dir = Vector3Normalize(Vector3Subtract(agujero, planeta));
        Vector3 perp = { -dir.z, 0, dir.x };
        if (Vector3Length(perp) < 0.01f) perp = { 0, 1, 0 };
        perp = Vector3Normalize(perp);

        Vector3 perpRot = {
            perp.x * cosf(ang) - perp.z * sinf(ang),
            sinf(ang),
            perp.x * sinf(ang) + perp.z * cosf(ang)
        };

        Vector3 pos;
        Vector3 vel;
        Color c;

        if (tipo == 0) {
            // TIPO 0: Flujo hacia agujero (como antes)
            float radio = 1.2f + GetRandomValue(0, 300) / 100.0f;
            float altura = GetRandomValue(-150, 150) / 100.0f;

            pos = Vector3Add(planeta,
                Vector3Add(
                    Vector3Scale(dir, 1.0f + GetRandomValue(0, 100) / 100.0f),
                    Vector3Add(
                        Vector3Scale(perpRot, radio),
                        { 0, 0, altura }
                    )
                )
            );

            float velMag = 6.0f + GetRandomValue(0, 400) / 100.0f;
            Vector3 velDir = Vector3Normalize(Vector3Add(dir, Vector3Scale(perpRot, 0.3f)));
            vel = Vector3Scale(velDir, velMag);

            unsigned char intensidad = (unsigned char)(150 + velMag * 10);
            c = { 255, intensidad, (unsigned char)(50 + velMag * 20), 255 };
        }
        else {
            // TIPO 1: Órbita alrededor del planeta (anti-horario)
            float radio = 2.0f + GetRandomValue(0, 400) / 100.0f; // 2.0 a 6.0
            float altura = GetRandomValue(-100, 100) / 100.0f;

            // Posición en órbita circular
            pos = Vector3Add(planeta,
                Vector3Add(
                    Vector3Scale(perpRot, radio),
                    { 0, 0, altura }
                )
            );

            // Velocidad tangencial PURA (anti-horaria)
            // v = sqrt(GM/r) para órbita circular
            float velOrbita = sqrtf(200.0f / radio) * (0.8f + GetRandomValue(0, 40) / 100.0f);

            // Dirección tangencial anti-horaria
            Vector3 tangencial = { -perpRot.y, perpRot.x, 0 };
            tangencial = Vector3Normalize(tangencial);

            // Añadir pequeña componente hacia agujero para que eventualmente caigan
            vel = Vector3Add(
                Vector3Scale(tangencial, velOrbita),
                Vector3Scale(dir, 1.5f) // Lento acercamiento al agujero
            );

            c = { 100, 255, 100, 255 }; // Verde claro
        }

        particulas[idx].reset(pos, vel, c, tipo);
    }

    void update() {
        calcularFocos();
        for (int i = 0; i < MAX_PARTICULAS; i++) {
            particulas[i].actualizar(planeta, agujero);
            if (particulas[i].debeReset(planeta, agujero)) spawn(i);
        }
    }

    void dibujarCuerpos() {
        // Planeta
        DrawSphere(planeta, 1.5f, Color{ 255, 200, 50, 255 });
        DrawSphereWires(planeta, 1.6f, 16, 16, ORANGE);

        for (int i = 0; i < 4; i++) {
            float ang = i * 90 * DEG2RAD + GetTime() * 2.0f;
            Vector3 r = { cosf(ang) * 1.8f, sinf(ang) * 1.8f, 0 };
            DrawLine3D(planeta, Vector3Add(planeta, r), Color{ 255, 150, 0, 150 });
        }

        // Agujero negro
        for (int i = 0; i < 5; i++) {
            float r = 1.0f + i * 0.8f;
            float alpha = 1.0f - i * 0.15f;
            Color c = Color{
                (unsigned char)(255 * alpha),
                (unsigned char)(100 * alpha),
                (unsigned char)(255 * alpha),
                (unsigned char)(150 * alpha)
            };
            DrawCylinderEx(
                Vector3Add(agujero, { 0, 0, -0.05f * (5 - i) }),
                Vector3Add(agujero, { 0, 0, 0.05f * (5 - i) }),
                r, r, 32, c
            );
        }

        for (int i = 0; i < 4; i++) {
            float ang = -i * 90 * DEG2RAD - GetTime() * 3.0f;
            Vector3 r = { cosf(ang) * 2.2f, sinf(ang) * 2.2f, 0 };
            DrawLine3D(agujero, Vector3Add(agujero, r), Color{ 200, 0, 255, 150 });
        }

        DrawSphere(agujero, 0.8f, BLACK);
        DrawSphereWires(agujero, 0.9f, 16, 16, PURPLE);

        DrawCylinderEx(
            Vector3Add(agujero, { 0,0,1 }),
            Vector3Add(agujero, { 0,0,12 }),
            0.3f, 0, 8, Color{ 150, 0, 255, 100 }
        );
        DrawCylinderEx(
            Vector3Add(agujero, { 0,0,-1 }),
            Vector3Add(agujero, { 0,0,-12 }),
            0.3f, 0, 8, Color{ 150, 0, 255, 100 }
        );
    }

    void dibujarParticulas() {
        for (auto& p : particulas) p.dibujar(agujero);
    }
};

// ============================================
// SLIDER
// ============================================
struct Slider {
    Rectangle r;
    float minV, maxV, * val;
    const char* label;
    Color col;

    Slider(float x, float y, float w, float h, float mn, float mx, float* v, const char* l, Color c)
        : r{ x,y,w,h }, minV(mn), maxV(mx), val(v), label(l), col(c) {
    }

    void update() {
        if (IsMouseButtonDown(MOUSE_LEFT_BUTTON)) {
            Vector2 m = GetMousePosition();
            if (CheckCollisionPointRec(m, r)) {
                float pct = (m.x - r.x) / r.width;
                *val = minV + pct * (maxV - minV);
            }
        }
        *val = fmaxf(minV, fminf(maxV, *val));
    }

    void draw() {
        DrawRectangleRec(r, DARKGRAY);
        float pct = (*val - minV) / (maxV - minV);
        DrawRectangle(r.x, r.y, r.width * pct, r.height, col);
        DrawRectangleLinesEx(r, 2, WHITE);
        DrawText(label, (int)r.x, (int)r.y - 18, 16, RAYWHITE);
        DrawText(TextFormat("%.2f", *val), (int)(r.x + r.width + 5), (int)r.y, 14, YELLOW);
    }
};

// ============================================
// MAIN
// ============================================
int main() {
    InitWindow(1280, 720, "Lorenz - Doble Comportamiento");
    SetTargetFPS(60);

    CamaraControl cam;
    Sistema sys;
    sys.init();

    Slider s1(20, 60, 180, 18, 1, 30, &sigma, "Sigma", ORANGE);
    Slider s2(20, 110, 180, 18, 1, 50, &rho, "Rho", GREEN);
    Slider s3(20, 160, 180, 18, 0.1f, 10, &beta, "Beta", BLUE);
    Slider s4(20, 210, 180, 18, 0.001f, 0.05f, &dt, "dt", RED);
    Slider s5(20, 260, 180, 18, 0.0f, 0.5f, &factorLorenz, "Lorenz", PURPLE);
    Slider s6(20, 310, 180, 18, 0.0f, 20.0f, &fuerzaSpin, "Spin", YELLOW);
    Slider* sliders[] = { &s1, &s2, &s3, &s4, &s5, &s6 };

    bool pause = false;

    while (!WindowShouldClose()) {
        cam.actualizar();
        for (auto s : sliders) s->update();
        if (IsKeyPressed(KEY_SPACE)) pause = !pause;
        if (IsKeyPressed(KEY_R)) sys.init();
        if (!pause) sys.update();

        BeginDrawing();
        ClearBackground(Color{ 2, 2, 10, 255 });

        BeginMode3D(cam.obtenerCamara());
        DrawGrid(60, 2.0f);
        sys.dibujarCuerpos();
        sys.dibujarParticulas();
        EndMode3D();

        DrawRectangle(10, 10, 280, 360, Color{ 0,0,0,180 });
        DrawText("LORENZ - DOBLE COMPORTAMIENTO", 20, 15, 20, RAYWHITE);
        DrawText("Naranja: Flujo a agujero", 20, 38, 12, ORANGE);
        DrawText("Verde: Orbita planeta (anti-horario)", 20, 52, 12, GREEN);
        DrawText("Arrastra: Rotar | Rueda: Zoom", 20, 70, 11, GRAY);

        for (auto s : sliders) s->draw();

        DrawText(TextFormat("FPS: %d %s", GetFPS(), pause ? "[PAUSA]" : ""), 20, 340, 14, pause ? RED : GREEN);

        DrawCircle(300, 30, 8, Color{ 255, 200, 50, 255 });
        DrawText("Estrella (anti-horario)", 315, 25, 12, WHITE);
        DrawCircle(300, 55, 8, PURPLE);
        DrawText("Agujero (horario)", 315, 50, 12, WHITE);
        DrawCircle(300, 80, 8, GREEN);
        DrawText("Orbitas", 315, 75, 12, WHITE);

        if (cam.arrastrando) DrawCircle(GetMouseX(), GetMouseY(), 8, Color{ 255, 255, 255, 80 });

        EndDrawing();
    }

    CloseWindow();
    return 0;
}
```

# Parámetros elegidos - El orden en el caos

| Parámetro         | Significado físico                             | Efecto visual                                      |
| ----------------- | ---------------------------------------------- | -------------------------------------------------- |
| **Sigma (σ)**     | Número de Prandtl - viscosidad térmica         | Controla la "apertura" de las alas del atractor    |
| **Rho (ρ)**       | Número de Rayleigh - diferencia de temperatura | Controla la intensidad del caos (ρ > 24.74 = caos) |
| **Beta (β)**      | Parámetro geométrico - relación de aspecto     | Controla la "aplanamiento" del atractor            |
| **dt**            | Paso de tiempo de integración                  | Precisión vs velocidad de simulación               |
| **Factor Lorenz** | Peso del campo de Lorenz vs gravedad           | Cuánto domina el caos sobre la física newtoniana   |
| **Spin**          | Intensidad de rotación diferencial             | Fuerza del "remolino" cerca de cada cuerpo         |

## Valores sugeridos para máximo enrollamiento

| Escenario                | Sigma | Rho  | Beta  | dt    | Lorenz | Spin | Descripción visual                            |
| ------------------------ | ----- | ---- | ----- | ----- | ------ | ---- | --------------------------------------------- |
| **Clásico Lorenz**       | 10.0  | 28.0 | 2.667 | 0.01  | 0.10   | 8.0  | Doble hélice estándar, enrollamiento moderado |
| **Caos intenso**         | 10.0  | 35.0 | 2.667 | 0.008 | 0.15   | 12.0 | Más vueltas, trayectorias impredecibles       |
| **Estructura ajustada**  | 14.0  | 28.0 | 2.0   | 0.01  | 0.08   | 10.0 | Atractor más "cerrado", enrollamiento denso   |
| **Máximo enrollamiento** | 10.0  | 28.0 | 8.0/3 | 0.005 | 0.20   | 15.0 | **Recomendado**: Muchas vueltas antes de caer |
| **Disco fino**           | 10.0  | 28.0 | 1.5   | 0.01  | 0.05   | 6.0  | Aplanado, menos vueltas, caída directa        |

## Relación matemática
**El atractor clásico funciona cuando:**<br>
$$\rho>\sigma \cdot \frac{\sigma+\beta+3}{\sigma-\beta-1}\approx 24.74$$
**Para** $\sigma=10,\beta=\frac{8}{3}$:
- $\rho=28 \rightarrow Caos-estandar$
- $\rho>30 \rightarrow Caos-mas-intenso, mas-vueltas$
- $\rho<24 \rightarrow, Sin-caos,-trayectoria-estable$
